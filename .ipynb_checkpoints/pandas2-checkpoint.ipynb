```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu, sigma= 2, 5
data= np.random.normal(mu, sigma, size=[20,5])
plt.hist(data, bins=5); #los metodos deben estas con letras minusculas
df= pd.DataFrame(data, columns=["v","w","x","y","z"])
plt.xticks(rotation=45)
```

```python
df[(df>0) & (df < 2)] ##mayores a 0  y menores a 2 & =and
```

```python
df= pd.DataFrame({
    "A":[7,8,-6,8],
    "B":[12,np.nan,1,np.nan],
    "C":[4,np.nan,np.nan,np.nan],
    "D":[4,np.nan,-2,10]})

df
```

```python
df.loc[4]=[5,8,12, -16] ## loc te ubica en la fila 4, y a esa fila se la agregas 
df
```

```python
df.isnull() ##mos dice donde tiene valores nan
```

```python
df.isna()
```

```python
df.isna().sum().sum() #Wel primer sum nos devuelve el nuemro de nulas en la primera fila, el segundo en todo el data frame
```

```python
df.dropna(axis=0, thresh=2, inplace=True) #axis, 0 para columnas 1 para filas, tresh define umbral
```

```python
df
```

```python
df.fillna(value=1) 
```

```python
df.fillna(value=df.mean()) ##los esta reemplazando con la media de esa columna
```

```python
df.mean(axis=1) ##media de las filas
#df.mean(axis=0)
```

```python
df["C"].fillna(value=df["C"].mean())
```

```python
df.loc[3].fillna(value=df.loc[3].min())
```
