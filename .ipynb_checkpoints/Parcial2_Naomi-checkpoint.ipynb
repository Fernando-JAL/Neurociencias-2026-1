```python
import mne
import numpy as np
import pandas as pd
from scipy.signal import coherence, hilbert
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import scipy.io
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
from matplotlib import colormaps
from networkx.algorithms.community import greedy_modularity_communities, modularity
```

```python
co_matrix_path = r"C:\Users\shade\OneDrive\Documents\GitHub\Neurociencias-2026-1\Coactivation_matrix.mat"
```

```python

coactivation_dict= scipy.io.loadmat(co_matrix_path)
coactivation_dict
```

```python
# mundo pequeño
matriz= coactivation_dict['Coactivation_matrix']
G_co = nx.from_numpy_array(matriz)
G_co = G_co.to_undirected()
# coeficiente de clustering medio (C)
C = nx.average_clustering(G_co, weight='weight')
L = nx.average_shortest_path_length(G_co, weight='weight')
# Crear red aleatoria equivalente (mismo número de nodos y densidad)
nodos = G_co.number_of_nodes()
aristas = G_co.number_of_edges()
p = (2 * aristas) / (nodos * (nodos - 1))  # probabilidad equivalente
G_rand = nx.gnp_random_graph(nodos, p)
C_rand = nx.average_clustering(G_rand)
L_rand = nx.average_shortest_path_length(G_rand)
# Coeficiente de mundo pequeño (σ)
sigma = (C / C_rand) / (L / L_rand)
print(f"C = {C:.3f}, L = {L:.3f}")
print(f"C_rand = {C_rand:.3f}, L_rand = {L_rand:.3f}")
print(f"Coeficiente de mundo pequeño (σ): {sigma:.3f}")
```

```python
#comunidades del grafo
matriz= coactivation_dict['Coactivation_matrix']
G_co = nx.from_numpy_array(matriz)
G_co = G_co.to_undirected()
communities = greedy_modularity_communities(G_co)
mod = modularity(G_co, communities)
print ('Communities=',communities,'Modularidad=', mod)
```

```python
print("Nodos:", list(G_co.nodes()))
```

```python
#  hubs
degree_centrality_co = nx.degree_centrality(G_co) 
sorted_degree_co = sorted(degree_centrality_co.items(), key=lambda x: x[1], reverse=True)
#  hubs principales
print("hubs por grado:")
for nodo, valor in sorted_degree_co[:5]:
    print(nodo, valor)
```

```python
# eficiencia global
eficiencia_global_co = nx.global_efficiency(G_co)
# eficiencia local
eficiencia_local_co = nx.local_efficiency(G_co)
print ('Eficiencia global=',eficiencia_global_co, 'Eficiencia local=', eficiencia_local_co)
```

```python
# Cálculo del grado por nodo (ordendos de mayor a menor)
df_co= pd.DataFrame(coactivation_dict['Coactivation_matrix'])
grafo_co= nx.from_pandas_adjacency(df_co, create_using=nx.DiGraph) 
grado_total_co= dict(grafo_co.degree(weight='weight'))
print ('grado total:',grado_total_co)
```

```python
#matrices de conectividad
#corr pearson
corr_matrix = np.corrcoef(coactivation_dict['Coactivation_matrix'])
corr_df_coact = pd.DataFrame(corr_matrix)
np.fill_diagonal(corr_df_coact.values, 0)

#coh espectral
fmin, fmax = 8, 13   
sfreq_target = 256   
coh_matrix_coact = np.zeros((638,638))
for i in range(638):
    for j in range(638):
        f, Cxy = coherence(coactivation_dict['Coactivation_matrix'][i], coactivation_dict['Coactivation_matrix'][j], fs=sfreq_target, nperseg=sfreq_target*2)
        mask = (f >= fmin) & (f <= fmax)
        coh_matrix_coact[i, j] = np.mean(Cxy[mask])
coh_df_coact = pd.DataFrame(coh_matrix_coact)
np.fill_diagonal(coh_df_coact.values, 0)

# plv
analytic_signal_coact = hilbert(coactivation_dict['Coactivation_matrix'])
phase_data_coact = np.angle(analytic_signal_coact)
plv_matrix_coact = np.zeros((638,638))
for i in range(638):
    for j in range(638):
        phase_diff_coact = phase_data_coact[i] - phase_data_coact[j]
        plv_matrix_coact[i, j] = np.abs(np.sum(np.exp(1j * phase_diff_coact)) / phase_diff_coact.size)
plv_df_coact = pd.DataFrame(plv_matrix_coact)
np.fill_diagonal(plv_df_coact.values, 0)
```

```python
edf_path = r"C:\Users\shade\OneDrive\Documents\GitHub\Neurociencias-2026-1\S03_datasets\EEG_practica\chb01_01.edf"
sfreq_target = 256        
raw = mne.io.read_raw_edf(edf_path, preload=True)
raw.resample(sfreq_target)
raw.filter(1., 45., fir_design='firwin')
data = raw.get_data()
ch_names = raw.info['ch_names']
n_channels = len(ch_names)
```

```python
analytic_signal_coact = hilbert(data)
phase_data_coact = np.angle(analytic_signal_coact)

plv_matrix_coact = np.zeros((n_channels, n_channels))
for i in range(n_channels):
    for j in range(n_channels):
        phase_diff_coact = phase_data_coact[i] - phase_data_coact[j]
        plv_matrix_coact[i, j] = np.abs(np.sum(np.exp(1j * phase_diff_coact)) / phase_diff_coact.size)

plv_df_coact = pd.DataFrame(plv_matrix_coact, index=ch_names, columns=ch_names)
np.fill_diagonal(plv_df_coact.values, 0)
```

```python
# Grafo 2D
grafo_coact = nx.from_pandas_adjacency(plv_df_coact > 0.1, create_using=nx.DiGraph)  # umbral para eliminar conexiones débiles
pos_coact = nx.spring_layout(grafo_coact)
plt.figure(figsize=(5, 5))
nx.draw_networkx(
    grafo_coact, pos_coact, with_labels=True,
    node_color='blue', arrows=True,
    node_size=400, font_size=8
)
plt.show()
```

```python
# coeficiente de mundo pequeño
G_coact = grafo_coact.to_undirected()  # convertir el grafo a no direccionado para los cálculos

#coef cluster medio
C_coact = nx.average_clustering(G_coact, weight='weight')

#l del camino mas corto
L_coact = nx.average_shortest_path_length(G_coact, weight='weight')

# red aleatoria equivalente
num_nodos_coact = G_coact.number_of_nodes()
num_aristas_coact = G_coact.number_of_edges()
p_coact = (2 * num_aristas_coact) / (num_nodos_coact * (num_nodos_coact - 1))  # densidad equivalente
G_rand_coact = nx.gnp_random_graph(num_nodos_coact, p_coact)

# clustering
C_rand_coact = nx.average_clustering(G_rand_coact)
L_rand_coact = nx.average_shortest_path_length(G_rand_coact)

# cofieciente de mundo pequeño
sigma_coact = (C_coact / C_rand_coact) / (L_coact / L_rand_coact)

print(f"C = {C_coact:.3f}, L = {L_coact:.3f}")
print(f"C_rand = {C_rand_coact:.3f}, L_rand = {L_rand_coact:.3f}")
print(f"Coeficiente de mundo pequeño (σ): {sigma_coact:.3f}")
```

```python
#comunidades
from networkx.algorithms.community import greedy_modularity_communities, modularity
communities_coact = greedy_modularity_communities(G_coact)
mod_coact = modularity(G_coact, communities_coact)
print('Communities =', communities_coact)
print('Modularidad =', mod_coact)
```

```python
# hubs
centralidad = nx.degree_centrality(G_coact) 
ordencen = sorted(centralidad.items(), key=lambda x: x[1], reverse=True)
print("hubs:")
for nodo, valor in ordencen[:5]:
    print(nodo, valor)
```
