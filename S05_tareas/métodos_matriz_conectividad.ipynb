# Métodos para constuir las relaciones entre nodos
Las matrices de conectividad, buscan mostrar las relaciones resultantes entre pares de nodos (regiones, células, núcleos) de mediciones con distintas técnicas de neuroimagen. 



# Correlación de Pearson
El índice de correlación de Pearson es una prueba estadísticaa que nos da el grado de relación lineal entro dos variables. El índice va de -1 a 1 y nos ofrece información sobre la intensidad de la correlación y la dirección.
En un contexto de conectividad neuronal, la correlación de Pearson no toma en cuenta las características de los valores de la señal (por ejemplo las freucencias de adquisición), es adecuado utilizar cuando se quiere ver activación global e inespecífica.


# Coherencia espectral
La coherencia espectral es una metodo de analísis de señales. Esta toma en cuenta las diferencias de fase entre canales, la amplitud y la temporalidad. Por lo tanto una matriz de coherencia espectral nos diría que tan sincronizados estan los pares de nodos registrados en una misma frecuencia. Un valor de 1 significa una relación perfecta entre nodos.
Es adecuado utilizarlo para construir matrices a partir de EEG o MEG, y si se quiere analizar la conectividad en un ancho de banda específica.

# PLV
El phase locking value por sus siglas en inglés analiza la sincronización de fase entre dos señales sin tomar en cuenta la amplitud.
Sus valores se encuentran ente 0 y 1.
Es adecuado utilizarlo para construir matrices que representen relaciones temporales en las fases de los nodos que se vean diferenciadas por muchas tareas.
