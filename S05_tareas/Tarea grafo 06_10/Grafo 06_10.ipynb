```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#cargar archivos
df = pd.read_excel("EEG.xlsx")
```

```python
df
```

```python
G = nx.Graph()
```

```python
G = nx.from_pandas_adjacency(df)
plt.figure(figsize = (6, 6))
```

```python
pos = nx.spring_layout(G, seed=42) 
nx.draw(G, pos, with_labels=True, node_size=1000, node_color="purple"); 
plt.figure(figsize = (6, 6))
```
