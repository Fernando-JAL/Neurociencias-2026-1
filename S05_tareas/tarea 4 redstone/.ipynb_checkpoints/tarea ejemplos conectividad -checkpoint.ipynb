# Correlación de Pearson 

Esto mide la relación lineal entre dos señales es decir si suben o bajan al mismo tiempo 

un ejemplo en las neurociencias de su aplicación sería el fMRI o la resonancia magnética funcional donde para una señal llamada BOLD (blood oxygen level dependent) se construye uan matriz de conectividad funcional en base a varias regiones implicadas.

Toma en cuenta a los nodos que son cada región del cerebro por ejemplo el Hipocampo y cada valor en la matriz que es la correlación entre dos regiones y coo resultado da una matriz simétrica (NxN).

un ejemplo más especifico sería para estudiar redes en reposo , comparar redes de pacientes depresivos y con Alzherimer.


# Coherencia Espectral 

Esta coherencia mide la cantidad de sincromía que hay entre dos señales en una banda de frecuencia espeficica como lo sería la alfa (8-12 Hz) o la Beta (13-30 Hz)

Es básicamente el valor cruzado de las transformadas de fourier que esto implica como dos áreas del cerebro oscilan juntas en una frecuencia dada, un ejemplo en las neurociencias sería el EEG o el MEG ( electroencefalograma y magento encefalograma) aquí cada valor representa la fuerza de acoplamiento entre los electrodos en una banda especifica.

Ejemplo específico : estudios para epilepsia o para el Parkinson evaluando las ondas en ambos casos .

# PLV ( phase locking value ) 

El PLV mide la constancia de la diferencia de fases entre dos señales a lo largo del teimpo 

Esta no depende de la amplitud, depende de la fase instantánea obtenida. esta sería obtenida con otra cosa que se llama transformada de Hilbert. El ejemplo en neurociencias sería en EEG o MEG , en estas técnicas de imagen es util para analizar la sincronía entre electrodos , canales o en las mismas regiones cerebrales.

el PLV cercano a 1 = que las fases están en perfecta sincronía lo que resulta en una alta coherencia de fase.

Ejemplo específico : estudios de memoria, percepción visual , auditiva o para patrones de sueño o alteraciones del sueño.
