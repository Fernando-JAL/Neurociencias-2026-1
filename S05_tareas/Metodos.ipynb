## Correlación de Pearson

La correlación de Pearson te permite cuantificar qué tan linealmente vinculadas están dos variables. Su valor oscila entre −1 y 1, indicando tanto la fuerza como la dirección de la relación

- Compara dos series de datos punto por punto

- Calcula si los cambios en una señal se corresponden con cambios similares en la otra

se usa cuando:
-cuando se buscan relaciones globales y simples

- Cuando no importa la frecuencia de la señal
  
## Coherencia espectral

 evalúa la relación entre canales considerando factores como fase, amplitud y variaciones temporales. Una matriz basada en coherencia espectral refleja el nivel de sincronía entre pares de nodos dentro de una misma banda de frecuencia. Un valor cercano a 1 indica una coincidencia casi perfecta. Este método es especialmente útil en datos provenientes de EEG o MEG.

- Descompone las señales en frecuencias (mediante transformada de Fourier)
se usa cuando:
Cuando el interés está en la actividad oscilatoria
  
## PLV (Phase Locking Value)
Mide el grado de sincronización de fase entre dos señales, sin tomar en cuenta su amplitud. Los valores que genera se sitúan entre 0 y 1, donde 1 señala una alineación de fase perfecta.
