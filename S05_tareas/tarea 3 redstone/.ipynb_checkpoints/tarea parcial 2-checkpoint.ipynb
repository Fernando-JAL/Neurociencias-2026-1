Cada ejercicio está representado por una función con su docstring que contiene la instrucción (no incluye soluciones). Implementa las funciones por tu cuenta.

```python
import pandas as pd
import numpy as np

# -----------------------------
# 1-4: Concatenación de DataFrames
# -----------------------------

def ex1_concat_same_structure():
    """Ejercicio 1"""
    df1 = pd.DataFrame({
        'Nombre': ['Pau', 'Meche', 'Sofi'],
        'Edad': [20, 19, 21]
    })
    df2 = pd.DataFrame({
        'Nombre': ['Arath', 'Susana', 'Alejandro'],
        'Edad': [20, 50, 80]
    })
    resultado = pd.concat([df1, df2], ignore_index=True)
    return resultado


def ex2_concat_with_month_column():
    """Ejercicio 2"""
    enero = pd.DataFrame({
        'Venta': [100, 150, 200],
        'Producto': ['A', 'B', 'C']
    })
    enero['Mes'] = 'Enero'

    febrero = pd.DataFrame({
        'Venta': [120, 130, 180],
        'Producto': ['A', 'B', 'C']
    })
    febrero['Mes'] = 'Febrero'

    total = pd.concat([enero, febrero], ignore_index=True)
    return total


def ex3_concat_diff_columns_fillna():
    """Ejercicio 3"""
    df1 = pd.DataFrame({
        'Producto': ['A', 'B', 'C'],
        'Ventas': [10, 20, 30]
    })
    df2 = pd.DataFrame({
        'Producto': ['A', 'B', 'C'],
        'Descuento': [5, 0, 2]
    })
    unido = pd.concat([df1, df2], ignore_index=True).fillna(0)
    return unido


def ex4_concat_three_groups():
    """Ejercicio 4"""
    grupo1 = pd.DataFrame({'Alumno': ['Monse', 'Javier', 'Ximena'], 'Nota': [8, 9, 7]})
    grupo2 = pd.DataFrame({'Alumno': ['Roy', 'Rodrigo', 'Omar'], 'Nota': [6, 9, 8]})
    grupo3 = pd.DataFrame({'Alumno': ['Mariana', 'Marta', 'Sofía'], 'Nota': [7, 8, 9]})
    todos = pd.concat([grupo1, grupo2, grupo3], ignore_index=True)
    return todos

# -----------------------------
# 5-8: Listas
# -----------------------------

def ex5_remove_duplicates_preserve_order(lst):
    """Ejercicio 5"""
    nueva = []
    for x in lst:
        if x not in nueva:
            nueva.append(x)
    return nueva


def ex6_sort_tuples_by_second_element(tuples_list):
    """Ejercicio 6"""
    return sorted(tuples_list, key=lambda x: x[1])


def ex7_first_n_multiples_of_7(n):
    """Ejercicio 7"""
    return [7 * i for i in range(1, n+1)]


def ex8_filter_palindromes(strings):
    """Ejercicio 8"""
    return [s for s in strings if s == s[::-1]]

# -----------------------------
# 9-12: Conjuntos
# -----------------------------

def ex9_set_intersection_and_difference(set_a, set_b):
    """Ejercicio 9"""
    inter = set_a & set_b
    diff = set_a - set_b
    return (inter, diff)


def ex10_is_subset(set_a, set_b):
    """Ejercicio 10"""
    return set_a.issubset(set_b)


def ex11_count_unique_elements(lst):
    """Ejercicio 11"""
    return len(set(lst))


def ex12_unique_words_from_phrases(phrases):
    """Ejercicio 12"""
    palabras = set()
    for frase in phrases:
        for palabra in frase.split():
            palabras.add(palabra.lower())
    return palabras

# -----------------------------
# 13-16: Tuplas
# -----------------------------

def ex13_oldest_person(tuple_list):
    """Ejercicio 13"""
    return max(tuple_list, key=lambda x: x[1])


def ex14_convert_list_of_lists_to_tuples(list_of_lists):
    """Ejercicio 14"""
    return [tuple(i) for i in list_of_lists]


def ex15_sum_of_tuple_abc(t):
    """Ejercicio 15"""
    a, b, c = t
    return a + b + c


def ex16_sort_product_price(prod_price_list):
    """Ejercicio 16"""
    return sorted(prod_price_list, key=lambda x: x[1])

# -----------------------------
# 17-20: Diccionarios
# -----------------------------

def ex17_char_frequency(s):
    """Ejercicio 17"""
    freq = {}
    for c in s:
        freq[c] = freq.get(c, 0) + 1
    return freq


def ex18_best_student(grades_dict):
    """Ejercicio 18"""
    mejor = max(grades_dict.items(), key=lambda x: x[1])
    return mejor


def ex19_invert_dictionary(d):
    """Ejercicio 19"""
    return {v: k for k, v in d.items()}


def ex20_merge_dicts_sum_values(d1, d2):
    """Ejercicio 20"""
    resultado = d1.copy()
    for k, v in d2.items():
        resultado[k] = resultado.get(k, 0) + v
    return resultado

# -----------------------------
# 21-24: NumPy
# -----------------------------

def ex21_numpy_array_set_fifth():
    """Ejercicio 21"""
    arr = np.zeros(10)
    arr[4] = 1
    return arr


def ex22_numpy_matrix_3x3():
    """Ejercicio 22"""
    return np.arange(9).reshape(3, 3)


def ex23_numpy_statistics():
    """Ejercicio 23"""
    datos = np.random.randint(0, 100, 20)
    media = np.mean(datos)
    mediana = np.median(datos)
    desv = np.std(datos)
    return (media, mediana, desv)


def ex24_numpy_clip_values(arr):
    """Ejercicio 24"""
    arr = np.array(arr)
    arr = np.clip(arr, None, 10)
    return arr

# -----------------------------
# 25-28: Pandas
# -----------------------------

def ex25_filter_rows_age_over_30():
    """Ejercicio 25"""
    df = pd.DataFrame({
        'Nombre': ['Sophie', 'Valeria', 'Sebastían', 'Jose'],
        'Edad': [28, 35, 40, 22],
        'Ciudad': ['CDMX', 'Puebla', 'Monterrey', 'Toluca']
    })
    return df[df['Edad'] > 30]


def ex26_load_dict_into_dataframe_and_sort():
    """Ejercicio 26"""
    datos = {'Nombre': ['María', 'Darió', 'Kassandra'], 'Puntaje': [80, 95, 90]}
    df = pd.DataFrame(datos)
    df = df.sort_values(by='Puntaje', ascending=False)
    return df


def ex27_groupby_mean_value():
    """Ejercicio 27"""
    df = pd.DataFrame({
        'Grupo': ['A', 'A', 'B', 'B', 'C'],
        'Valor': [10, 20, 30, 40, 50]
    })
    return df.groupby('Grupo', as_index=False)['Valor'].mean()


def ex28_most_sold_product_from_sales_df():
    """Ejercicio 28"""
    ventas = pd.DataFrame({
        'Producto': ['A', 'B', 'A', 'C', 'B', 'A'],
        'Cantidad': [10, 5, 8, 3, 7, 9]
    })
    total = ventas.groupby('Producto')['Cantidad'].sum()
    mas_vendido = total.idxmax()
    return (mas_vendido, total[mas_vendido])

# -----------------------------
# 29-32: Lambdas
# -----------------------------

def ex29_map_lambda_cube(lst):
    """Ejercicio 29"""
    return list(map(lambda x: x**3, lst))


def ex30_filter_lambda_even(lst):
    """Ejercicio 30"""
    return list(filter(lambda x: x % 2 == 0, lst))


def ex31_sort_list_of_dicts_by_key(lst, keyname):
    """Ejercicio 31"""
    return sorted(lst, key=lambda x: x[keyname])


def ex32_pandas_apply_lambda_uppercase(names_list):
    """Ejercicio 32"""
    df = pd.DataFrame({'Nombre': names_list})
    df['MAYUS'] = df['Nombre'].apply(lambda x: x.upper())
    return df
```
