# Diferencia entre conectividad estructural y funcional
### Conectividad Estructural
Se refiere a las conexiones físicas que existen entre estructuras, en el caso del cerebro se suele referir a las conexiones que forman los tractos de sustancia blanca
### Conectividad Funcional
Se refiere a cómo se activa, actúan distintas áreas del cerebro aunque no estén físicamente cerca. Nos permite ver la activación de distintas areas en respuesta a una tarea o estado cognitivo
