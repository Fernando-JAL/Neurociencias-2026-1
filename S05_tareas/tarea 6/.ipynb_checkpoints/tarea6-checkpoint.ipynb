```python
excel_coord.set_index('Canal',inplace=True)
```

```python
for node in newG.nodes():
    if node in excel_coord.index:
        x, y = excel_coord.loc[node, ["x", "y"]]
        newG.nodes[node]["pos"] = (x, y)
```

```python
posicion = {n: newG.nodes[n]["pos"] for n in newG.nodes()}
```
