```python
#Crear un dataframe de 200 filas
#Con las columnas: tipos_sujetos, condición, region, neurotransmisor;
#Estas columnas deben tener una distribución uniforme 
#agregar una columna llamada 'presición' que tenga una distribución normal con mu=0.92, sigma=0.08
#Este Dataset pretende simular los resultados de un estudio en el que se miden variables cognitivas y 
#fisiológicas en diferentes regiones cerebrales, sujetos y condiciones experimentales
```

```python
import numpy as np
import pandas as pd

# Número de filas
n = 200  

# Definimos las opciones para cada columna categórica
tipos_sujetos = ["sujeto_A", "sujeto_B", "sujeto_C", "sujeto_D"]
condiciones = ["control", "experimental"]
regiones = ["hipocampo", "corteza", "amígdala", "tálamo"]
neurotransmisores = ["dopamina", "serotonina", "GABA", "glutamato"]

# Elegimos valores al azar de forma uniforme
col_tipos = np.random.choice(tipos_sujetos, size=n)
col_cond = np.random.choice(condiciones, size=n)
col_region = np.random.choice(regiones, size=n)
col_neuro = np.random.choice(neurotransmisores, size=n)

# Columna precisión con distribución normal
mu = 0.92
sigma = 0.08
col_precision = np.random.normal(mu, sigma, size=n)

# Crear el DataFrame
df = pd.DataFrame({
    "tipos_sujetos": col_tipos,
    "condición": col_cond,
    "región": col_region,
    "neurotransmisor": col_neuro,
    "precisión": col_precision
})

# Mostrar las primeras filas
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Histograma de conteo de sujetos
df["tipos_sujetos"].value_counts().plot(kind="bar")
plt.title("Conteo de datos por sujeto")
plt.xlabel("Sujetos")
plt.ylabel("Cantidad")
plt.show()

# Histograma de proporción de neurotransmisores
df["neurotransmisor"].value_counts(normalize=True).plot(kind="bar")
plt.title("Proporción de neurotransmisores")
plt.xlabel("Neurotransmisor")
plt.ylabel("Proporción")
plt.show()
```

```python
# Proporción de condiciones por sujeto
proporciones = df.groupby("tipos_sujetos")["condición"].value_counts(normalize=True)
print("\nProporciones de condiciones por sujeto (groupby):")
print(proporciones)

# Máx, mín, media de precisión por sujeto
stats = df.groupby("tipos_sujetos")["precisión"].agg(["max", "min", "mean"])
print("\nEstadísticas de precisión por sujeto (groupby):")
print(stats)

# Histograma de regiones por condición
df.groupby("condición")["región"].value_counts().unstack().plot(kind="bar", stacked=True)
plt.title("Histograma de regiones activadas por condición")
plt.xlabel("Condición")
plt.ylabel("Frecuencia")
plt.show()
```
