- hacerlo por filtrado de dataframes y usando groupby

### Groupby

```python
# por medio de groupby
df_sujetos_cond = (df.groupby(["sujeto", "condicion"]))
df_sujetos_cond.describe()
```

```python
df_sujetos_prec = df.groupby("sujeto") # al ser los únicos valores numéricos se calcula para la precisión.
df_sujetos_prec.describe()
```

```python
df_cond_region = df.groupby(["condicion", "region"]).size().reset_index(name="conteo")
df_cond_region
```

```python
for cond, subdf in df_cond_region.groupby("condicion"):
    plt.bar(subdf["region"], subdf["conteo"])
    plt.title(f"Regiones activadas en condición {cond}")
    plt.xlabel("Región")
    plt.ylabel("Frecuencia")
    plt.show()
```
