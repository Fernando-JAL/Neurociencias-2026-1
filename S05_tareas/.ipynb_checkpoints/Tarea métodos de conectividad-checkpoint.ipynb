# Tarea

Correlación de Pearson
- Se utiliza como una medida estadística de dos variables aleatorias cuantitativas y se da a conocer la dirección y la intensidad de relación que hay entre ellas.
- En este caso para la matriz de conectividad se miden la relación lineal entre dos señales de tiempo
- Se utiliza para la conectividad estática funcional en la resonacia magnética y electroencefalografía
- En el caso para el electroencefalograma, medir la apmplitud entre dos electrodos a lo largo del tiempo
- Cuando son valores cercanos a 1, quiere decir que hay una relación lineal entre dos señales, -1 inidican una relación inversa y 0 cuando hay ausencia de esas señales


Coherencia espectral:
- Es una correlación que hay entre dos señales una frecuencia específica
- Se utiliza para medir la relación entre dos señales en una banda de frecuencia específica, combinando información de amplitud y fase.
- Se utiliza para la conectividad dependiente de frecuencia en EEG y MEG.
- Permite analizar la sincronización en bandas específicas (delta, theta, alpha, beta, gamma).
- Los valores cercanos a 1 indican alta sincronización en esa frecuencia, valores cercanos a 0 indican sincronización ausente.

Phase Locking Value (PLV):
- Se utiliza para medir la consistencia relativa de la fase entre dos señales a lo largo del tiempo o ensayos.
- Mide cómo dos señales mantienen su fase sincronizada.
- Se utiliza para la conectividad basada en fase, ignorando la amplitud de la señal.
- Igual es muy útil para el EEG y MEG para estudios de eventos o ensayos repetidos.
- Los valores cercanos a 1 indican fase perfectamente sincronizada, valores cercanos a 0 indican fase aleatoria.
