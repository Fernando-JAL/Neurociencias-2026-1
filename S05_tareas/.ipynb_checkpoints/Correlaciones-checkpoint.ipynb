## Correlacion de Pearson

Se usa en variables cuantitativas, es un  índice  que  mide  el grado de covariación entre distintas variables relacionadas linealmente (ya que puede haber variables fuertemente relacionadas pero no de forma lineal, y no se aplicaría esta correlación, por ejemplo, la relación entre la  ansiedad y  el rendimiento). 
Sus valores absolutos oscilan entre –1 y +1.  En el primer caso la relación es perfecta positiva y en el segundo perfecta negativa y si el valor del coeficiente de correlación es nulo o cercano a 0, quiere decir que las variables prácticamente no están relacionadas linealmente. Sin embargo, correlación no implica causalidad. 

## Coherencia espectral

Es una medida utilizada para analizar la relación entre dos señales en el dominio de la frecuencia. Específicamente, mide qué tan bien una señal puede predecirse a partir de otra en diferentes frecuencias. En otras palabras, mide relación lineal entre amplitud y fase, como función de la frecuencia. La coherencia se calcula como el cuadrado de la magnitud de la densidad espectral cruzada normalizada por las densidades espectrales de potencia individuales de las señales. Puede ser útil en EEG para estudiar conectividad funcional, evaluar cambios en sincronización en trastornos neurológicos, medir acoplamiento entre regiones durante una tarea cognitiva.

## PHASE LOCKING VALUE (PLV)

Es una medida de la consistencia de la diferencia de fase entre dos señales oscilatorias a lo largo del tiempo. Es útil en EEG para evaluar conectividad. Si dos señales tienen una diferencia de fase estable, están “phase-locking”, es decir, sus ciclos ocurren con sincronía.
Si la diferencia de fase cambia mucho, la sincronización es baja. 
1 → sincronización perfecta de fase.
0 → no hay sincronización (las fases están aleatorias).
