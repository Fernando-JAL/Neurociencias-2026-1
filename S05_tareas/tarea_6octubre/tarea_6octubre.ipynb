Crear un grafo bidireccional con las dos columnas de la primera hoja y plotear grafo.

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Leer el archivo .xlsx
df = pd.read_excel('EEG_3D_coordinates.xlsx')
# Definir coordenadas x y y
cor_x = df['x']
cor_y = df['y']
# Configurar el tamaño del gráfico
plt.figure(figsize=(4, 4))
# Graficar los puntos
plt.scatter(cor_x, cor_y, color='blue', label='Puntos');
```
