```python
def cuenta_atras(n):
    if n >0:##esto evita q la función sea infinita
        print(n)
        cuenta_atras(n-1)
    else:
        print("boomboommm")
    print("fin de la función", n)

cuenta_atras(3)      

```

```python
def factorial (n):
    if n==1 or n==0:
       return (1)
    else:
        return n*factorial(n-1)
```

```python

factorial(13)
```

```python
#sumatoria
def sumatoria(n):
    if n == 0:
        return 0
    else:
        return 5**n + sumatoria(n-1)

print(sumatoria(1))
```
