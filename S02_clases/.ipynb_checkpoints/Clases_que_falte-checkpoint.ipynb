```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import openpyxl
import numpy as np
```

```python
## los archivos que se van a usar
excel_conectividad = r"C:\Users\monce\OneDrive\Documentos\GitHub\Neurociencias-2026-1\EEG.xlsx"
excel_coordenadas = r"C:\Users\monce\OneDrive\Documentos\GitHub\Neurociencias-2026-1\S03_datasets\EEG_3D_coordinates.xlsx"
```

```python
## leer los archivos como diccionario el sheet name es para que lea todas las hojas del excel
conectividad_dict = pd.read_excel(excel_conectividad, sheet_name=None)

coordenadas_dict = pd.read_excel(excel_coordenadas, sheet_name=None)
```

```python
conectividad_dict
```

```python
## especifica que hoja vamos a leer [0]
motor_df= conectividad_dict[list(conectividad_dict.keys())[0]] #[list(conectividad_dict.keys()) esto espec√≠fica que se va 
## convertir las llaves de un diccionariario en una lista y iteras solo el elemento 0. Las llaves de ese diccionario son las hojas
coordenadas_df= coordenadas_dict[list(coordenadas_dict.keys())[0]]
```

```python
motor_df
```

```python
### redefine el aindice para que sea mas facil convertirlo a grafo
## inplace reemplaza el dataframe original
motor_df.set_index("Unnamed: 0", inplace=True, drop=True) ## Reemplza los indices, por unnamed jiji
coordenadas_df.set_index("Canal", inplace=True, drop=True)
```

```python
motor_df
```

```python
## dibujar el grafo
plt.figure(figsize=(5,5))
```

```python
G = nx.from_pandas_adjacency(motor_df)
nx.draw_circular(G, with_labels=True)
```

```python
# PLOTEO 2D
### crear un dicconario de posiciones
pos2D= {canal: (coordenadas_df.loc[canal, 'x'],
                coordenadas_df.loc[canal, 'y']) for canal in coordenadas_df.index}
pos2D ##Este ciclo for, extrae las coordenadas x y y del data frame Coordenadas_df, por cada canal x y y, crean un diccionario 
## de su canal y de las coordenadas
```

```python
plt.figure(figsize=(4,4))
nx.draw_networkx(G, pos=pos2D) ##pos permite indicar las coordanas que establecimos.
plt.show() 
```

```python
pos3D= {canal: (coordenadas_df.loc[canal, 'x'],
                coordenadas_df.loc[canal, 'y'],
                coordenadas_df.loc[canal, 'z']) for canal in coordenadas_df.index}

fig = plt.figure(figsize=(4,4))
```

```python
pos3D= {canal: (coordenadas_df.loc[canal, 'x'],
                coordenadas_df.loc[canal, 'y'],
                coordenadas_df.loc[canal, 'z']) for canal in coordenadas_df.index}

fig = plt.figure(figsize=(4,4))

## definir el tama√±o
ax = fig.add_subplot(111, projection='3d')
## contar los subplots que voy a tener y que de esos vaya uno por uno (111)

## dibujar nodos en 3d
for canal, (x, y, z) in pos3D.items():
    ax.scatter(x, y, z, s=100)
    ax.text(x, y, z+0.03, canal, fontsize=10, ha='center') ##asign puntos a las coordenadas

for i, j, data in G.edges(data=True):
    x= [pos3D[i][0], pos3D[j][0]]
    y= [pos3D[i][1], pos3D[j][1]]
    z= [pos3D[i][2], pos3D[j][2]]
    ax.plot(x, y, z, linewidth=data['weight']*5) ## esta parte dibuja arista

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Red de conectividad EEG (3D)')
plt.show()
```
