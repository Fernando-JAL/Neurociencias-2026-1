```python
import numpy as np
import pandas as pd
import networkx as nx

import seaborn as sns
import matplotlib.pyplot as plt
import scipy.io as sp

file_path= r"C:\Users\monce\OneDrive\Documentos\GitHub\Neurociencias-2026-1\S03_datasets\BCT\cat.mat"

from scipy.io import loadmat

file_dict = loadmat(file_path) ## Convertir a diccionario
```

```python
matriz= file_dict["CIJctx"] #Vector de 52
columnas= file_dict["Names"] ##52x52

cat_df= pd.DataFrame(data= matriz, columns= columnas, index= columnas)
cat_df

plt.figure(figsize=(4,4))
ax= sns.heatmap(cat_df.values,
                annot= False, cmap ="hot", fmt=".2f",
                xticklabels=cat_df.columns, yticklabels= cat_df.index)
ax.set(xlabel="c", ylabel="channels")
plt.show() 
```
