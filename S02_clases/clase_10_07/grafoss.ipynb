```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import networkx as nx
```

```python
excel_conectividad  = r'C:\Users\Juan\Documents\GitHub\Neurociencias-2026-1\S03_datasets\EEG.xlsx'
```

```python
excel_coordenadas = r"C:\Users\Juan\Documents\GitHub\Neurociencias-2026-1\S03_datasets\EEG_3D_coordinates.xlsx"
```

```python
conectividad_dict = pd.read_excel(excel_conectividad, sheet_name = None)
coordenadas_dict = pd.read_excel(excel_coordenadas, sheet_name = None)
```

```python
motordf = conectividad_dict[list(conectividad_dict.keys())[0]]
motordf
```

```python
coord_df = coordenadas_dict[list(coordenadas_dict.keys())[0]]
coord_df
```

```python
#Redefinir los Ã­ndices

motordf.set_index('Unnamed: 0', drop=True, inplace=True)
coord_df.set_index('Canal', inplace =True, drop = True)
```

```python
plt.figure(figsize = (5,5)) #Hace una figura de un tamaÃ±o indicado
G = nx.from_pandas_adjacency(motordf) #Crea una matriz de adyacencia a partir del dataframe
nx.draw_circular(G, with_labels=True) #Hace el grafo de la matriz de adyacencia, manteniendo 
#plt.show()
```

```python
#Grafo 2D
pos2D = {canal: (coord_df.loc[canal, 'x'],
                coord_df.loc[canal, 'y']) for canal in coord_df.index}

plt.figure(figsize = (5,5)) #Hace una figura de un tamaÃ±o indicado
nx.draw_networkx(G, with_labels=True, pos=pos2D) #Hace el grafo de la matriz de adyacencia, manteniendo 
#plt.show()
```

```python
#Grafo 3D
pos3D = {canal: (coord_df.loc[canal, 'x'],
                coord_df.loc[canal, 'y'],
                 coord_df.loc[canal, 'z']) for canal in coord_df.index}
fig = plt.figure(figsize = (5,5))
ax = fig.add_subplot(111, projection = '3d') #el 111 indica la malla de subplots que va a crear, una de 111 solo serÃ¡ un subplot
for canal,(x,y,z) in pos3D.items():
    ax.scatter(x,y,z, s=100)
    ax.text(x,y,z+0.03, canal, fontsize = 10, ha='center')

for i,j, data in G.edges(data=True):
    x = [pos3D[i][0], pos3D[j][0]]
    y = [pos3D[i][1], pos3D[j][1]]
    z = [pos3D[i][2], pos3D[j][2]]
    ax.plot(x,y,z,linewidth = data['weight']*5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Red de conectividad')
plt.show()
```
