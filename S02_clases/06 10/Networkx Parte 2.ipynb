```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G= nx.Graph()
G.add_nodes_from([('A', {'pos': [3,2]}),
                  ('B', {'pos': [2, -5]}),
                  ('C', {'pos': [5, 1]}),
                  ('D', {'pos': [-1, 1]}),
                  ('E', {'pos': [5, -4]}), ] )
G.add_edges_from([('D', 'A', {'weight':1.3}),
                  ('A', 'C',  {'weight':3.9}),
                  ('D', 'E',  {'weight':0.8}),
                  ('B', 'C',  {'weight':0.6}),
                  ('A', 'E',  {'weight':2.2}),
                  ('D', 'B',  {'weight':3.4})])
plt.figure(figsize=(2, 2))
```

```python
positions= nx.get_node_attributes(G, 'pos')
nx.draw_networkx(G, pos=positions, node_size=50, font_size=6, width=3,
                 node_color='pink', edge_color='green')
# Agregar los pesos a las aristas
weights = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos=positions, edge_labels=weights);
#nx.draw_networkx_edges(G,edgellist=G. edges())
```

```python
nx.to_pandas_edgelist(G)
list(G.adjacency())
adjacency=nx.to_pandas_adjacency(G)
adjacency
```

```python
adjacency.to_csv('adjacency.csv')
# Guardad un archivo en excel
```

```python
import pandas as pd
df=pd.read_csv('adjacency.csv')
df.set_index('Unnamed: 0', drop=True, inplace=True)
newG = nx.from_pandas_adjacency(df)
```
