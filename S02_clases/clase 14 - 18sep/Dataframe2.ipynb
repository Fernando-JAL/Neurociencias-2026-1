```python
df = pd.DataFrame({
    "A": [7, 8, -6, 8],
    "B": [12, np.nan, 1, np.nan],
    "C":[4, np.nan, np.nan, np.nan],
    "D": [ 4, np.nan, -2, -10]})
df
```

```python
df.loc[4]= [5, 8, 12, -7]
df
```

```python
df.isna() # van a aparecer los NaN
```

```python
df.isna(), df.isnull()
```

```python
df.isna().sum().sum()  #cuenta los trues
```

```python
df
```

```python
df.dropna(axis=0, thresh=2, inplace=True) #Va a eliminar los Na en toda la fila y el axis es para la columna y thresh permite que no se eliminen todos los NaN para que no altere todo el data frame
```

```python
df
```

```python
df.fillna(value="X") #llenar el NaN con otra cosa escrita
```

```python
df.fillna(value=df.mean()) #llenar el espacio con el valor medias por columnas (solamente se puede por columnas)
```

```python
df.mean(axis=1)  #aquí es en filas
```

```python
df["C"].fillna(value=df["C"].mean())  #agarrar solo los valores de C
```

```python
df.loc[3].fillna(value=df.loc[3].mean()) #agarrar solo los valores de la fila 3
```
