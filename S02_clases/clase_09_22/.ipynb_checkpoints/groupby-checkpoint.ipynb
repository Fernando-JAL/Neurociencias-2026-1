```python
# Para cada sujeto, sacar la proporción del tipo de condición 
#Para cada sujeto, calcular: máx, min y media de su precisión 
# para cada condición hacer el histograma de cada región activada. 
#Hacerlo mediante 2 métodos: filtrado de data frame y usando groupby
```

```python
#1
```

```python
df2 = df[df['sujetos']==sujetos]
proporciones = df2['condición2'].value_counts(normalize=True)
```

```python
display(proporciones)
```

```python
sujetos = df['sujetos'].unique()

for sujeto in sujetos:
    total = df[df['sujetos'] == sujeto].shape[0]
    condiciones = df[df['sujetos'] == sujeto]['condición2'].value_counts()
    proporciones = condiciones / total
```

```python
proporciones 
```

```python
proporciones = df.groupby(['sujetos', 'condición2']).size().unstack(fill_value=0)
proporciones = proporciones.div(proporciones.sum(axis=1), axis=0)
```

```python
proporciones
```

```python
#2
```

```python
for sujeto in df['sujetos'].unique():
    precisiones = df[df['sujetos'] == sujeto]['precisión'].explode()
    print(f"{sujeto} → Máx: {precisiones.max():.2f}, Mín: {precisiones.min():.2f}, Media: {precisiones.mean():.2f}")
```

```python
calculos = df.groupby('sujetos')['precisión'].agg(['max', 'min', 'mean'])
```

```python
calculos
```

```python
import matplotlib.pyplot as plt

condiciones = df['condición2'].unique()

for cond in condiciones:
    regiones = df[df['condición2'] == cond]['region2']
    plt.figure()
    regiones.value_counts().plot(kind='bar')
    plt.show()
```

```python
region_counts = df.groupby(['condición2', 'region2']).size().unstack(fill_value=0)

region_counts.T.plot(kind='bar', subplots=True, layout=(len(region_counts.columns), 1), figsize=(8, 6), legend=False)
plt.tight_layout()
plt.show()
```
