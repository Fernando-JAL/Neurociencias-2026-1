```python
tipos_sujetos = ['S' + str(i).zfill(2) for i in range(4)]
generos = ['M', 'F']
condicion = ['çontrol', 'experimental']
region = ['hipocampo', 'cortex prefrontal', 'amigdala']
neurotransmisores = ['dopamina', 'serotonina'] 

# crea un dataframe de 200 filas 
# con las columnas : tipos_sujetos, generos,condicion,region, neurotrasmisor
#estas columnas deben tener distribución normal 
#agregar una columna llamada "precisión" que tenga una distribución normal mu= 0.92, sigma = 0.08
#este dataset pretende simular los resultados de un estudio en el que se miden variables cognitivas
#y fisiológicas en difernetes regiones cerebrales, sujetos y condicoones experimentales
```

```python
# Número de muestras
n = 200

# Función para muestreo normal sobre categorías
def sample_normal_categorical(categories, size):
    # Asignar probabilidades normales centradas en el medio
    x = np.linspace(-3, 3, len(categories))
    probs = np.exp(-x**2 / 2)
    probs /= probs.sum()
    return np.random.choice(categories, size=size, p=probs)

# Crear el DataFrame
df = pd.DataFrame({
    'tipos_sujetos': sample_normal_categorical(tipos_sujetos, n),
    'generos': sample_normal_categorical(generos, n),
    'condicion': sample_normal_categorical(condicion, n),
    'region': sample_normal_categorical(region, n),
    'neurotransmisor': sample_normal_categorical(neurotransmisores, n),
    'precisión': np.random.normal(loc=0.92, scale=0.08, size=n)
})

# Visualizar las primeras filas
print(df.head())
```

```python
import numpy as np
import pandas as pd

# Número de muestras
n = 200

# Listas de categorías para los sujetos
tipos_sujetos = ['control', 'experimental', 'placebo', 'activo']
generos = ['masculino', 'femenino', 'no binario']
condicion = ['reposo', 'estimulación', 'estrés']
region = ['frontal', 'parietal', 'occipital', 'temporal']
neurotransmisores = ['dopamina', 'serotonina', 'noradrenalina', 'acetilcolina']
grupo_edad = ['joven', 'adulto', 'mayor']

# Función para muestreo normal sobre categorías
def sample_normal_categorical(categories, size):
    x = np.linspace(-3, 3, len(categories))
    probs = np.exp(-x**2 / 2)
    probs /= probs.sum()
    return np.random.choice(categories, size=size, p=probs)

# Crear el DataFrame con 6 columnas de sujetos y 6 de precisión
df = pd.DataFrame({
    'tipos_sujetos': sample_normal_categorical(tipos_sujetos, n),
    'generos': sample_normal_categorical(generos, n),
    'condicion': sample_normal_categorical(condicion, n),
    'region': sample_normal_categorical(region, n),
    'neurotransmisor': sample_normal_categorical(neurotransmisores, n),
    'grupo_edad': sample_normal_categorical(grupo_edad, n),
    'precisión_1': np.random.normal(loc=0.92, scale=0.08, size=n),
    'precisión_2': np.random.normal(loc=0.90, scale=0.07, size=n),
    'precisión_3': np.random.normal(loc=0.88, scale=0.09, size=n),
    'precisión_4': np.random.normal(loc=0.91, scale=0.06, size=n),
    'precisión_5': np.random.normal(loc=0.93, scale=0.05, size=n),
    'precisión_6': np.random.normal(loc=0.89, scale=0.1, size=n)
})

# Visualizar las primeras filas
print(df.head())
```
