```python
import numpy as np
import pandas as pd
import networkx as nx
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.io 
file_path= r"C:\Users\Luz Mar√≠a\OneDrive\Escritorio\Neurociencias-2026-1\S03_datasets\BCT\macaque47.mat"
```

```python
file_dict= scipy.io.loadmat(file_path)
```

```python
file_dict
```

```python
conect_df=file_dict['CIJ']
nombres= file_dict['Names']
```

```python
np.size(nombres)
```

```python
macaco_df= pd.DataFrame(data=conect_df, columns=nombres, index=nombres)
```

```python
plt.figure(figsize=(4,4))
ax= sns.heatmap(macaco_df.values,
                annot=False, cmap='hot', fmt='.2f',
                xticklabels= macaco_df.columns, yticklabels=macaco_df.index)
ax.set(xlabel='channels', ylabel='channels')
plt.show()
```

```python
#actividad
macaquito= macaco_df.iloc[15:35,15:35]
plt.figure(figsize=(4,4))
ax= sns.heatmap(macaquito.values,
                annot=False, cmap='hot', fmt='.2f',
                xticklabels= macaquito.columns, yticklabels=macaquito.index)
ax.set(xlabel='channels', ylabel='channels')
plt.show()
garfo = nx.from_pandas_adjacency(macaquito, create_using=nx.DiGraph)
pos = nx.spring_layout(garfo)
plt.figure(figsize=(5,5))
nx.draw_networkx(garfo, pos, with_labels=True, 
                 node_color='pink', arrows=True, 
                 node_size=400, font_size=8)
plt.show()
grado_total = dict(garfo.degree(weight=None))
grado_total
```

```python
# act
#actualizar el main para ver el archivo macaco csv
#plotear en 3d el grafo completo 
#reemplazar el .excel por .csv
path_macaca= r"C:\Users\Luz Mar√≠a\OneDrive\Escritorio\Neurociencias-2026-1\S03_datasets\BCT\macaco_M132_F99_LH.csv"
```

```python
macaca_df= pd.read_csv(path_macaco)
macaca_df.set_index('region',inplace=True, drop=True)
macaca_df.index
```

```python
G=nx.from_pandas_adjacency(macaco_df, create_using=nx.DiGraph());
pos3D= {region:(macaca_df.loc[region, 'x'],
                macaca_df.loc[region, 'y'],
                macaca_df.loc[region, 'z']) for region in macaca_df.index}
fig= plt.figure(figsize=(5,5))
ax= fig.add_subplot(111, projection='3d')
for region, (x,y,z) in pos3D.items():
    ax.scatter(x,y,z, s=100)
    ax.text(x,y,z+0.03, region, fontsize=10, ha='center')
for i, j, data in G.edges(data=True):
    x= [pos3D[i][0], pos3D[j][0]]
    y= [pos3D[i][1], pos3D[j][1]]
    z= [pos3D[i][2], pos3D[j][2]]
    ax.plot(x, y, z, linewidth= data['weight']*5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Conectividad de macaco')
plt.show()
```
