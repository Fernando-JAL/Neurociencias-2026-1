```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns 
import scipy.io
```

```python
#Definimos la ruta del archivo.mat
file_path = r"C:\Users\ecere\Documents\GitHub\Neurociencias-2026-1\S03_datasets\BCT\macaque47.mat"
#abrir el contenido del archivo .mat
file_dict = scipy.io.loadmat(file_path)
```

```python
file_dict
```

```python
matriz = file_dict['CIJ']
columnas = file_dict['Names']
```

```python
matriz.shape, len(columnas)
```

```python
macaco_df = pd.DataFrame(data = matriz, columns = columnas, index= columnas)
```

```python
macaco_df.columns = [columnas.split(sep=' ')[0] for columnas in macaco_df.columns]
macaco_df.index = macaco_df.columns
```

```python
ax = sns.heatmap(macaco_df, annot=False, fmt=".0f", cmap='hot',
                xticklabels=macaco_df.columns, yticklabels=macaco_df.columns)
plt.show()
```

```python
#Mapa de calor
ax = sns.heatmap(macaco_df_15_35, annot=False, fmt=".0f", cmap='hot',
                xticklabels=macaco_df_15_35.columns, yticklabels=macaco_df_15_35.columns)
plt.show()
```

```python
#Grafo dirigido
G=nx.from_pandas_adjacency( macaco_df.iloc[15:36, 15:36] , create_using=nx.DiGraph())

pos=nx.spring_layout(G)
nx.draw_networkx(G,with_labels=True, arrows=True, arrowstyle="-|>", connectionstyle ="arc3,rad=0.15")
```

```python
#Calcular grado de grafo 
# Grado total (entrada + salida)
grado_total = dict(G.degree())

# Grado de entrada
grado_entrada = dict(G.in_degree())

# Grado de salida
grado_salida = dict(G.out_degree())
```

```python
grados_df = pd.DataFrame({
    "Grado total": grado_total,
    "Grado de entrada": grado_entrada,
    "Grado de salida": grado_salida
})
print(grados_df)
```
