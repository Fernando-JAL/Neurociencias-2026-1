```python
names = [n.strip() for n in data['Names'].flatten()] 
CIJctx = data['CIJctx']
```

```python
df = pd.DataFrame(CIJctx, index=names, columns=names)
print("DataFrame creado correctamente:", df.shape)
display(df.head()) # con esto nos aseguramos que este bien hecho el dataframe y debe darnos 52 y 52
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df, cmap='Purples')# si quieres cambiar a un color más aesthetic buscar en google cmap python 
plt.title('Mapa de calor - CIJctx (Completo)')
plt.show()
```

```python
valores = df.values.flatten()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(valores, bins=20, kde=False)
plt.title('Histograma de valores - CIJctx')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
val_counts = pd.Series(valores).value_counts(normalize=True) * 100
print("Porcentaje de cada valor:")
print(val_counts.sort_index())
```

```python
filters = [0, 1, 2]
filtered_dfs = {}

for f in filters:
    filtered = df.copy()
    filtered[filtered <= f] = 0
    filtered_dfs[f] = filtered
    
    total = df.size
    eliminados = (df.values <= f).sum()
    porcentaje_eliminado = (eliminados / total) * 100

    print(f"\nFiltro <= {f}: {porcentaje_eliminado:.2f}% de datos eliminados")
```

```python
for f in filters:
    plt.figure(figsize=(10,8))
    sns.heatmap(filtered_dfs[f], cmap='Oranges')
    plt.title(f'Mapa de calor - Filtro <= {f}')
    plt.show()
```

```python
for f in filters:
    G = nx.from_numpy_array(filtered_dfs[f].values)
    G = nx.relabel_nodes(G, dict(enumerate(names)))

    plt.figure(figsize=(8,6))
    nx.draw_circular(G, with_labels=True, node_size=200, font_size=8)
    plt.title(f'Grafo 2D - Filtro <= {f}')
    plt.show()
```
