```python
# Definimos la ruta del archivo .mat
file_path = r"C:\Users\OMEN CI7\Documents\repository\Neurociencias-2026-1\S03_datasets\BCT\cat.mat"

# Leemos el contenido del archivo .mat
file_dict = scipy.io.loadmat(file_path)
```

```python
file_dict
```

```python
matriz = file_dict['CIJctx']
columnas = file_dict['Names']
```

```python
df = pd.DataFrame(data=matriz, columns=columnas, index=columnas)
```

```python
plt.figure(figsize=(2, 2))
ax = sns.heatmap(df.values,
                 annot=False, cmap='hot', fmt=".2f",
                 xticklabels=df.columns, yticklabels=df.index)
ax.set(xlabel="channels", ylabel="channels")
plt.show()
```
