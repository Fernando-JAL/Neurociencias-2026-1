```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
#agregar nodos --> Pos= posiciones 
G.add_nodes_from([("A",{"pos":[3,3]}),
                  ("B",{"pos":[2,-5]}),
                  ("C",{"pos":[5,2]}),
                  ("D",{"pos":[-1,2]}),
                  ("E",{"pos":[4,-5]}) ]),
```

```python
G.add_edges_from([("D", "A", {"weight" : 1.3}), 
                  ("A", "C", {"weight" : 3.9}), 
                  ("D", "E", {"weight" : 0.8}), 
                  ("B", "C", {"weight" : 0.6}), 
                  ("A", "E", {"weight" : 2.2}), 
                  ("B", "D", {"weight" : 3.4})])
```

```python
plt.figure(figsize = (2, 2))
nx.draw_networkx(G)
```

```python
plt.figure(figsize = (2, 2))
```

```python
positions = nx.get_node_attributes(G, "pos") 
weights = nx.get_edge_attributes(G, "weight") 
```

```python
nx.draw_networkx(G, pos = positions) 
nx.draw_networkx_edge_labels(G, pos = positions, edge_labels = weight); 
plt.figure(figsize = (2, 2))
```

```python
#agregar ms caracteristicas 
nx.draw_networkx(G, pos = positions, node_size = 100, font_size = 6, width = 3, 
                 node_color = "purple", edge_color = "pink")
nx.draw_networkx_edge_labels(G, pos = positions, edge_labels = weights); 
nx.to_pandas_edgelist(G)
list(G.adjacency())
```

```python
#Crear matriz de adyacencia 
adjacency = nx.to_pandas_adjacency(G)
adjacency
adjacency.to_csv("adjacency_ej.csv")
```

```python
import pandas as pd
# podemos cargar archivos
df = pd.read_csv("adjacency_ej.csv")
```
