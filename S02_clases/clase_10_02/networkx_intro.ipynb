```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
G.nodes(), G.edges()
```

```python
G.add_nodes_from([ 'A', 'B', 'C', 'D', 'E' ])
```

```python
G.nodes(), G.edges()
```

```python
G.add_edges_from( [ ('B', 'C'), ('C', 'A'), ('D', 'E') ] )
```

```python
G.add_edge("C", "D")

nx.draw_networkx(G)
```
