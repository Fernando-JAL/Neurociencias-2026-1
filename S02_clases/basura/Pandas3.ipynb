```python
enteros=np.random.randint(0,5,size=200)
sujetos=["s"+str(i).zfill(2) for i in enteros]
sujetos 
```

```python
genero=np.random.randint(0,2,size=200)
generos= []
for valor in genero:
    if valor==0:
        generos.append("hombre")
    else:
        generos.append("mujer")
        
```

```python
condición=np.random.randint(0,2,size=200)
condiciones=[]

for valor in condición:
    if valor==0:
        condiciones.append("control")
    else:
        condiciones.append("experimental")
```

```python
region=np.random.randint(0,3,size=200)
regiones=[]
for valor in region:
    if valor==0:
        regiones.append("hipocampo")
    elif valor == 1:
        regiones.append("cx prefrontal")
    else:
        regiones.append("amigdala")
```

```python
condición=np.random.randint(0,2,size=200)
neurotransmisores=[]

for valor in condición:
    if valor==0:
        neurotransmisores.append("dopamina")
    else:
        neurotransmisores.append("serotonina")
```

```python
df = pd.DataFrame({
    "Sujetos": sujetos,
    "Genero": generos,
    "Condicion": condiciones,
    "Region": regiones,
    "Neurotransmisor": neurotransmisores,
})
df ## al parecer se pueden hacer data drames apartir de bibliotecas
```

```python
#agregar una columna llamada, "precisión" con una distribución normal con mu=0.92, sigma=0.08
df["precisión"]= np.random.normal(0.92, 0.08, size=[200,1])
df
```
