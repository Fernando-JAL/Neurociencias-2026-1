```python
df3= pd.DataFrame({
    "A":[7,8,-6,8],
    "B":[12,np.nan,1,np.nan],
    "C":[4,np.nan,np.nan,np.nan],
    "D":[4,np.nan,-2,10]})

df3
```

```python
## como agregar una fila .loc[donde va estar] = [ lo que le quieres agregar]
df3.loc[10]=[5,8,12,-7]
df3
```

```python
## identificar valores que no estan .isnua(), df.isnull()
df3.isna()
```

```python
## contar los falsos devolvio numero de trues de cada una de las columnas (cuantos no hay)
df3.isna().sum()
```

```python
df3.isna().sum().sum ## se lo vuelvo aplicar, me da pues de todo el data frame todos los null
df3.isna().sum().sum()
```

```python
df3
```

```python
##quitar donde no hay .dropna() --> quita filas 
                            ##argumentos que hay ( axis= 0 para filas, 1 = para columnas )
df3.dropna()
```

```python
df3.dropna(axis=1)
```

```python
df3.dropna(axis=0)
```

```python
df3.dropna(axis=0, thresh=2) ## umbral de cuantos naN vas a permitir 
```

```python
##sobre escribe el dataframe original (inplace=True)
df3.dropna(axis=0, thresh=2, inplace=True)
df3
```

```python
###.fillna( se remplaza naN con alguna otra cosa) --> remplax
df3.fillna(value="olis")
```

```python
df3.fillna(value=df3.mean()) ##el axis por default se va por las columnas 
```

```python
##medias por filas
df3.mean(axis=1)
```

```python
df3["C"].fillna(df3["C"].mean())
```
