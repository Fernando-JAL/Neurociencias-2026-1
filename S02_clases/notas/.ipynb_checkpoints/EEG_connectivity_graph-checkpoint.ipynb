
## 3) Funciones para calcular conectividad por banda

Se incluye una función que usa `mne.connectivity.spectral_connectivity` para calcular conectividad por bandas (coherencia, PLV, wPLI, correlation, etc.). Ajusta `method` y `fmin/fmax` según necesites.


````python

def compute_connectivity(raw, picks, sfreq=None, method='coh', fmin=8., fmax=12., tmin=None, tmax=None, mt_adaptive=False):
    """Calcula una matriz de conectividad entre las señales seleccionadas.
    - raw: objeto mne Raw
    - picks: lista de nombres de canales (ordenará según picks)
    - method: 'coh', 'wpli', 'pli', 'plv', 'corr', 'coh'
    - fmin,fmax: banda de frecuencia (Hz)
    """
    # if sfreq is None:
    #     sfreq = raw.info['sfreq']
    # # seleccionar datos
    # picks_idx = mne.pick_channels(raw.ch_names, include=picks)
    # data, times = raw[picks_idx, :]
    # # spectral_connectivity espera shape (n_signals, n_times) en general
    # con, freqs, times, n_epochs, n_tapers = spectral_connectivity(
    #     data[np.newaxis, ...], method=method, sfreq=sfreq,
    #     fmin=fmin, fmax=fmax, faverage=True, verbose='ERROR'
    # )
    # # con tiene forma (n_signals, n_signals) si faverage=True
    # mat = con[:, :, 0] if con.ndim==3 else con
    # # asegurar matriz simétrica (para métodos simétricos)
    # mat = np.squeeze(mat)
    ch_names = raw.info['ch_names']
    data = raw.get_data()
    n_channels = len(ch_names)

    coh_matrix = np.zeros((n_channels, n_channels))
    for i in range(n_channels):
        for j in range(n_channels):
            f, Cxy = coherence(data[i], data[j], fs=sfreq, nperseg=sfreq*2)
            mask = (f >= fmin) & (f <= fmax)
            coh_matrix[i, j] = np.mean(Cxy[mask])
    
    mat = pd.DataFrame(coh_matrix, index=ch_names, columns=ch_names)
    return mat

# Ejemplo de uso (no ejecutado hasta correr la celda):
# picks = BIPOLAR_CHANNELS  # si tus canales coinciden
# mat_alpha = compute_connectivity(raw, picks, method='coh', fmin=8., fmax=12.)
````


## 4) Construir grafos y calcular métricas

Se incluyen funciones para:
- construir grafo a partir de matriz de conectividad (con umbral o porcentaje de enlaces)
- calcular métricas globales (densidad, eficiencia, clustering, small-worldness)
- métricas nodales (grado, betweenness, centralidad propia)


````python

def mat_to_graph(mat, node_names=None, threshold='percent', thresh_val=0.2, directed=False):
    """Convierte una matriz de conectividad en un grafo de networkx.
    - threshold: 'percent' (mantener top X%), 'value' (usar umbral absoluto), or 'none' (usar completa)
    - thresh_val: si 'percent', valor entre 0-1 que indica fracción de aristas a mantener;
                 si 'value', umbral absoluto.
    """
    if node_names is None:
        node_names = list(range(mat.shape[0]))
    # symmetrizar
    W = np.array(mat.copy())
    # Si queremos grafos no dirigidos, promediamos con su transpuesta
    if not directed:
        W = (W + W.T) / 2.0
    # enmascarar NaNs
    W = np.nan_to_num(W)
    # aplicar umbral
    if threshold == 'percent':
        # conservar top k porcentuales de valores (excluyendo diagonal)
        n = W.shape[0]
        tri_idx = np.triu_indices(n, k=1)
        vals = np.abs(W[tri_idx])
        kth = np.quantile(vals, 1-thresh_val)
        mask = np.abs(W) >= kth
        W_thresh = W * mask
    elif threshold == 'value':
        mask = np.abs(W) >= thresh_val
        W_thresh = W * mask
    else:
        W_thresh = W.copy()
    # construir grafo
    if directed:
        G = nx.from_numpy_array(W_thresh, create_using=nx.DiGraph)
    else:
        G = nx.from_numpy_array(W_thresh)
    mapping = {i: node_names[i] for i in range(len(node_names))}
    G = nx.relabel_nodes(G, mapping)
    # asignar pesos
    for u, v, d in G.edges(data=True):
        d['weight'] = W_thresh[node_names.index(u), node_names.index(v)] if isinstance(node_names[0], str) else W_thresh[u, v]
    return G

def graph_metrics(G):
    metrics = {}
    metrics['n_nodes'] = G.number_of_nodes()
    metrics['n_edges'] = G.number_of_edges()
    metrics['density'] = nx.density(G)
    try:
        metrics['avg_clustering'] = nx.average_clustering(G, weight='weight')
    except Exception:
        metrics['avg_clustering'] = nx.average_clustering(G)
    try:
        metrics['avg_shortest_path'] = nx.average_shortest_path_length(G, weight='weight')
    except Exception:
        metrics['avg_shortest_path'] = float('nan')
    # eficiencia global (NetworkX no tiene función directa; implementamos)
    def global_efficiency(G):
        lengths = dict(nx.all_pairs_dijkstra_path_length(G, weight='weight'))
        n = G.number_of_nodes()
        s = 0.0
        for u in G.nodes():
            for v in G.nodes():
                if u == v: 
                    continue
                try:
                    d = lengths[u][v]
                    s += 1.0 / d if d>0 else 0.0
                except KeyError:
                    s += 0.0
        return s / (n*(n-1))
    try:
        metrics['global_efficiency'] = global_efficiency(G)
    except Exception:
        metrics['global_efficiency'] = float('nan')
    # nodal medidas
    metrics['degree'] = dict(G.degree(weight='weight'))
    metrics['betweenness'] = nx.betweenness_centrality(G, weight='weight', normalized=True)
    metrics['eigenvector'] = nx.eigenvector_centrality_numpy(G, weight='weight')
    return metrics
````


## 5) Detección de comunidades y hubs
Se incluye código para detectar comunidades con Louvain (si está instalado) o con `greedy_modularity_communities` como alternativa.


````python

def detect_communities(G):
    if community_louvain is not None:
        # convertir a grafo no dirigido y sin pesos negativos
        G_copy = G.copy()
        for u,v,d in G_copy.edges(data=True):
            if 'weight' in d and d['weight'] < 0:
                G_copy[u][v]['weight'] = abs(d['weight'])
        partition = community_louvain.best_partition(G_copy, weight='weight')
        # partition: dict nodo->comunidad
        return partition
    else:
        # usar greedy modularity (devuelve sets de nodos)
        communities = list(nx.algorithms.community.greedy_modularity_communities(G, weight='weight'))
        partition = {}
        for i, com in enumerate(communities):
            for n in com:
                partition[n] = i
        return partition

def plot_graph(G, coords_dict=None, node_size_map=None, title='Grafo EEG'):
    plt.figure(figsize=(8,6))
    if coords_dict is not None:
        pos = coords_dict
    else:
        pos = nx.spring_layout(G)
    sizes = 200
    if node_size_map is not None:
        # normalizar tamaños
        vals = np.array(list(node_size_map.values()), dtype=float)
        # vals = minmax_scale(vals, feature_range=(50,800))
        sizes = [vals[i] for i in range(len(vals))]
        mapping = {n: sizes[i] for i,n in enumerate(node_size_map.keys())}
        sizes = [mapping[n] for n in G.nodes()]
    nx.draw_networkx_nodes(G, pos, node_size=sizes)
    nx.draw_networkx_edges(G, pos, alpha=0.4)
    nx.draw_networkx_labels(G, pos, font_size=8)
    plt.title(title)
    plt.axis('off')
    plt.show()
````


## 6) Dinámica temporal (sliding windows)

Ejemplo de cómo dividir la señal en ventanas y calcular una serie temporal de matrices de conectividad. Luego puedes analizar cómo cambian la modularidad o la centralidad a lo largo del tiempo.


````python

def sliding_window_connectivity(raw, picks, window_sec=5.0, step_sec=1.0, method='coh', fmin=8., fmax=12.):
    sfreq = raw.info['sfreq']
    win_samp = int(window_sec * sfreq)
    step_samp = int(step_sec * sfreq)
    picks_idx = mne.pick_channels(raw.ch_names, include=picks)
    data, times = raw[picks_idx, :]
    n_samples = data.shape[1]
    mats = []
    timestamps = []
    for start in range(0, n_samples - win_samp + 1, step_samp):
        segment = data[:, start:start+win_samp]
        # spectral_connectivity espera 3D (n_epochs, n_signals, n_times) -> usamos 1 epoch
        con, freqs, times_c, n_epochs, n_tapers = spectral_connectivity(
            segment[np.newaxis, ...], method=method, sfreq=sfreq, fmin=fmin, fmax=fmax, faverage=True, verbose='ERROR'
        )
        mat = np.squeeze(con)
        if mat.ndim == 3:
            mat = mat[:, :, 0]
        mats.append(mat)
        timestamps.append(start / sfreq)
    return np.array(mats), np.array(timestamps)

# Ejemplo de llamada:
# mats, times = sliding_window_connectivity(raw, picks=BIPOLAR_CHANNELS, window_sec=5, step_sec=1)
````
