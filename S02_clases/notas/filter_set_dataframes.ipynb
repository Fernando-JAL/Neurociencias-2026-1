```python
import numpy as np
import pandas as pd
```

```python
df_list = []
n_data = 5

for idx in range(n_data):
    df = pd.DataFrame(np.random.randint(10, size=[3, 3]))
    df_list.append(df)
```

```python
threshold, percentaje = 5, 0.6

# Convierte todos a numpy y apílalos
stacked = np.stack([df.values for df in df_list])      # shape = (N, filas, columnas)
# Matriz booleana: True donde valor > X
above = stacked > threshold                                # shape = (N, filas, columnas)

N = len(df_list)

# Cuenta cuántos superan el umbral
count_above = above.sum(axis=0)                    # shape = (filas, columnas)

# Condición del 60%
min_required = int(np.ceil(percentaje * N))

# Matriz final booleana
result_bool = count_above >= min_required

# Si quieres un dataframe:
result_df = pd.DataFrame(result_bool, index=df_list[0].index, columns=df_list[0].columns).astype(int)
```

```python
# Matriz donde los que no cumplen se ponen como NaN
masked = np.where(above, stacked, np.nan)

# Promedio ignorando NaN
mean_matrix = np.nanmean(masked, axis=0)

mean_df = pd.DataFrame(mean_matrix, index=df_list[0].index, columns=df_list[0].columns)
```
