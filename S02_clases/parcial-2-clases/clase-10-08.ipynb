```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Grafo 
G = nx.Graph()
G.add_nodes_from(['A', 'B', 'C', 'D', 'E'])
G.add_edges_from([('A', 'B'), 
                  ('A', 'D'), 
                  ('B', 'C'), 
                  ('C', 'E'), 
                  ('D', 'E')])
plt.figure(figsize = (2, 2))
nx.draw_circular(G) # Grafo no dirigido
```

```python
nx.draw_networkx(G, arrows = True, arrowstyle = '->') # Grafo dirigido
```

```python
# DataFrame que tenga weight'

# Nodos del grafo
index = ['A', 'B', 'C', 'D', 'E']

# Matriz (DataFrame) simtrica
matrix = pd.DataFrame(columns = index, index = index)

# Asignar pesos a las aristas (direccionalmente)
matrix.iloc[0, 1] = 2.1
matrix.iloc[1, 0] = 8.6
matrix.iloc[0, 3] = 7.5
matrix.iloc[3, 0] = 6.6
matrix.iloc[3, 4] = 4.2
matrix.iloc[4, 3] = 0.8
matrix.iloc[4, 2] = 3.56
matrix.iloc[2, 4] = 1.1
matrix.iloc[1, 2] = 10.2
matrix.iloc[2, 1] = 7.3

# Reemplazar valores NaN (vacos) por 0
matrix.fillna(0, inplace = True)

# Grafo dirigido a partir de la matriz 
G = nx.from_pandas_adjacency(matrix, create_using = nx.DiGraph())
plt.figure(figsize = (9, 5))
pos = nx.spring_layout(G) # Calcular la posicin de los nodos
nx.draw_networkx(G, pos, with_labels = True, arrows = True, arrowstyle = '->',  connectionstyle='arc3,rad=0.3')
weights = nx.get_edge_attributes(G, 'weight') # Extraer los pesos de las aristas desde el grafo
nx.draw_networkx_edge_labels(G, pos, edge_labels = weights, label_pos=0.3) # Dibujar valores de los pesos sobre las aristas
plt.show()
```
