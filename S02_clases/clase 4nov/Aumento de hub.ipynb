```python
#Mundo pequeño
coh_matrix = np.zeros((n_channels, n_channels))

for i in range(n_channels):
    for j in range(n_channels):
        f, Cxy = coherence(chb01_data[i], chb01_data[j], fs=sfreq_target, nperseg=sfreq_target*2)
        mask = (f >= fmin) & (f <= fmax)
        coh_matrix[i, j] = np.mean(Cxy[mask])
```

```python
G = nx.from_numpy_array(coh_matrix)
clust_coeff = nx.average_clustering(G, weight="weight")
```

```python
try:
    path_length = nx.average_shortest_path_length(G, weight="weight")
except nx.NetworkXError:
    path_length = np.nan  
G_rand = nx.gnm_random_graph(n=G.number_of_nodes(), m=G.number_of_edges())
clust_rand = nx.average_clustering(G_rand)
path_rand = nx.average_shortest_path_length(G_rand)
small_world_sigma = (clust_coeff / clust_rand) / (path_length / path_rand)
(path_length, clust_coeff, clust_rand, path_rand)
print("Coeficiente de mundo pequeño:", small_world_sigma)
```
