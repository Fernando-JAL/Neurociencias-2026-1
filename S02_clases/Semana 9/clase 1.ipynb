```python
import networkx as nx
```

```python
G= nx.Graph()
```

```python
G.add_nodes_from([('A',{'pos':[3,3]}),
                  ('B',{'pos':[2,-5]}),
                  ('C',{'pos':[5,2]}),
                  ('D',{'pos':[-1,2]}),
                  ('E',{'pos':[4,-5]})]);
```

```python
G.add_edges_from([('D','A',{'weight':1.3}),
                  ('A','C',{'weight':3.9}),
                  ('D','E',{'weight':0.8}),
                  ('B','C',{'weight':0.6}),
                  ('A','E',{'weight':2.2}),
                  ('D','B',{'weight':3.4})])
```

```python
nx.draw_shell(G)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(2,2))
# agregando posiciones a los nodos
position=nx.get_node_attributes(G,'pos')

nx.draw_networkx(G,pos=position,node_size=150, font_size=6, width=6, edge_color='r', node_color='pink')


#agregando los pesos a las aristas
weights=nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos=position, edge_labels=weights);
```

```python
edgelist=nx.to_pandas_edgelist(G)
```

```python
#list(G.adjacency())
adjancency=nx.to_pandas_adjacency(G)
```
