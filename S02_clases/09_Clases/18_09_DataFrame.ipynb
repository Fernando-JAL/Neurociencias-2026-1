```python
df = pd.DataFrame({
    'A':[7, 8, -6, 8],
    'B':[12, np.nan, 1, np.nan],
    'C':[4, np.nan, np.nan, np.nan],
    'D':[4, np.nan, -2, -10]})
df
```

```python
# para agregar filas

df.loc[4] = [5, 8, 12, -7]
df
```

```python
df.loc[10] = [5, 8, 12, -7]
df
```

```python
# para identificar los NaN
df.isnull()
```

```python
# o como alternativa:
df.isna()
```

```python
#si uso el sum me devulva para de las filas el numero del nulls
df.isna().sum()
```

```python
# para que me de el numero de nulls en todo el dataframe
df.isna().sum().sum()
```

```python
# no muestra las filas que tienen NaN
df.dropna()
```

```python
# para columnas
df.dropna(axis=1)
```

```python
# para filas
df.dropna(axis=0)
```

```python
# si tiene mas de ese numero marcado el thresh, quitalo
df.dropna(axis=0, thresh=2)
```

```python
# nos muestra el dataframe original
df
```

```python
df.dropna(axis=0, thresh=2, inplace=True)
```
