```python
import mne
import numpy as np
import pandas as pd
from scipy.signal import coherence, hilbert
import networkx as nx
import matplotlib. pyplot as plt
import scipy.io 
import seaborn as sns
from scipy.io import loadmat
from matplotlib import cm
```

```python
#  dataset de chb01_01.edf:
chb01=r"C:\Users\ecere\Documents\GitHub\Neurociencias-2026-1\S03_datasets\EEG_practica\chb01_01.edf"
sfreq_target = 256           
fmin, fmax = 8, 13 
raw = mne.io.read_raw_edf(chb01, preload=True)
raw.resample(sfreq_target)
raw.filter(1., 45., fir_design='firwin')
data = raw.get_data()
ch_names = raw.info['ch_names']
n_channels = len(ch_names)
```

```python
#Mundo pequeño 
corr_mat = np.corrcoef(data)
threshold = np.percentile(corr_mat, 50)
adj_matrix = np.where(corr_mat >= threshold, 1, 0)
G = nx.from_numpy_array(adj_matrix)
C = nx.average_clustering(G)
try:
    L = nx.average_shortest_path_length(G)
except nx.NetworkXError:
    # si el grafo no es conexo, tomar camino promedio de la componente gigante
    largest_cc = max(nx.connected_components(G), key=len)
    G_lcc = G.subgraph(largest_cc)
    L = nx.average_shortest_path_length(G_lcc)
    G_rand = nx.gnm_random_graph(n_channels, G.number_of_edges())
C_rand = nx.average_clustering(G_rand)
try:
    L_rand = nx.average_shortest_path_length(G_rand)
except nx.NetworkXError:
    largest_cc_rand = max(nx.connected_components(G_rand), key=len)
    G_lcc_rand = G_rand.subgraph(largest_cc_rand)
    L_rand = nx.average_shortest_path_length(G_lcc_rand)
    sigma = (C / C_rand) / (L / L_rand)

print(f"Coeficiente de clustering C: {C:.4f}")
print(f"Camino promedio L: {L:.4f}")
print(f"C_aleatorio: {C_rand:.4f}, L_aleatorio: {L_rand:.4f}")
print(f"mundo pequeño σ: {sigma:.4f}")
```

```python

# --- Obtener datos EEG ---
data = raw.get_data()  # canales x tiempos
n_canales = data.shape[0]
print("Número de canales:", n_canales)

# --- Matriz de conectividad y grafo ---
matriz_conect = np.corrcoef(data)
G = nx.from_numpy_array(matriz_conect)

# --- Comunidad única (todos los nodos juntos) ---
comunidades = [list(G.nodes())]
print(f"Comunidades encontradas: {len(comunidades)}")
print(f"Comunidad única: {comunidades[0]}")

# --- Calcular hubs (top 5 por grado) ---
grados = dict(G.degree())
sorted_degree = sorted(grados.items(), key=lambda x: x[1], reverse=True)
hubs = [n for n, _ in sorted_degree[:5]]
print("Hubs:", hubs)

# --- Tamaño de nodos proporcional al grado ---
max_grado = max(grados.values())
tamanos_nodos = {n: (grados[n] / max_grado) * 800 for n in G.nodes()}

# --- Posiciones 3D aleatorias ---
np.random.seed(42)
pos = {i: (np.random.rand(), np.random.rand(), np.random.rand()) for i in G.nodes()}

# --- Graficar ---
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot(111, projection='3d')

# Dibujar aristas
color_aristas = 'brown'
for i, j in G.edges():
    x = [pos[i][0], pos[j][0]]
    y = [pos[i][1], pos[j][1]]
    z = [pos[i][2], pos[j][2]]
    weight = matriz_conect[i, j]
    ax.plot(x, y, z, c=color_aristas, alpha=0.5)
    

# Dibujar nodos
for n in G.nodes():
    x, y, z = pos[n]
    size = tamanos_nodos[n]
    if n in hubs:
        ax.scatter(x, y, z, s=size*2.5, c='gold', edgecolors='black', linewidth=1.5,
                   marker='o', label='Hub' if 'Hub' not in ax.get_legend_handles_labels()[1] else "")
    else:
        ax.scatter(x, y, z, s=size, color=color_nodos[n], edgecolors='k')

# Título y leyenda
ax.set_title("Grafo 3D EEG - Comunidad única y hubs resaltados", fontsize=13)
plt.legend()
plt.show()

```
