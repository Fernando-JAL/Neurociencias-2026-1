```python
df= pd.DataFrame({
    'A':[7,8,-6,8],
    'B':[12,np.nan, 1, np.nan],
    'C':[4,np.nan, np.nan, np.nan],
    'D':[4,np.nan, -2,-10]})
df
```

```python
df.loc[4] = [5,8,12,-7]
df
```

```python
## nos dice donde tenemos valores nan
df.isna(), df.isnull()
```

```python
## cuentas los na?
df.isna().sum().sum()
```

```python
## quitas los nan, drop na, deshace las filas de na
df.dropna()
```

```python
## es para filas, para si quitar nan pero siendo mas accesibles
df.dropna(axis=0)
```

```python
df.dropna(axis=0, thresh=2)
```

```python
#regresa el data frame original, a menos de que pongamos inplace
df
```
