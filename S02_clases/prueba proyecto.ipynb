```python
# Filtrado de archivos .csv para obtener time_stimon de cada estímulo por sujeto

import pandas as pd
import glob
import os
import re

stimname_variables = ["apple_01b.jpg", "apple_02s.jpg", "apple_03s.jpg", "apple_04s.jpg", 
                      "avocado_01b.jpg", "avocado_02s.jpg", "avocado_03s.jpg", "avocado_04s.jpg", 
                      "banana_01b.jpg", "banana_02s.jpg", "banana_03s.jpg", "banana_04s.jpg", 
                      "beet_01b.jpg", "beet_02s.jpg", "beet_03s.jpg", "beet_04s.jpg",
                      "cheesecake_01s.jpg", "cheesecake_02s.jpg", "cheesecake_03s.jpg", "cheesecake_04s.jpg", 
                      "cherry_01b.jpg", "cherry_02s.jpg", "cherry_03s.jpg", "cherry_04n.jpg",
                      "hamburger_01s.jpg", "hamburger_02s.jpg", "hamburger_03s.jpg", "hamburger_04s.jpg",
                      "hotdog_01b.jpg", "hotdog_02s.jpg", "hotdog_03s.jpg", "hotdog_04s.jpg",
                      "lasagna_01b.jpg", "lasagna_02s.jpg", "lasagna_03s.jpg", "lasagna_04s.jpg",
                      "mango_01b.jpg", "mango_02s.jpg", "mango_03s.jpg", "mango_04s.jpg"

]

csv_files = glob.glob(r"C:\Users\marub\OneDrive\Escritorio\Datos proyecto modelos I\Eventos\sub-*_task-rsvp_events.csv")

resultados = []

for file_path in csv_files:
    try:
        df = pd.read_csv(file_path)

        # Extraer el nombre del archivo (ej: "sub-01_task-rsvp_events.csv")
        filename = os.path.basename(file_path)

        # Usar regex para sacar "sub-XX"
        match = re.search(r"(sub-\d+)", filename)
        subject_id = match.group(1) if match else "unknown"

        # Filtrar alimentos
        alimentos_df = df[df['stimname'].isin(stimname_variables)]

        # Seleccionar columnas
        selected = alimentos_df[['stimname', 'time_stimon']].copy()

        # Redondear
        selected['time_stimon'] = selected['time_stimon'].round(2)

        # Añadir columna del sujeto
        selected['subject'] = subject_id

        # Guardar
        resultados.append(selected)

        print(f"Processed {file_path}: Found {len(selected)} entries for {subject_id}.")

    except Exception as e:
        print(f"Error processing {file_path}: {e}")

# Concatenar
alimentos_df = pd.concat(resultados, ignore_index=True)

print("\nTotal rows collected:", len(alimentos_df))
print(alimentos_df.head())

# Crear columna numérica del sujeto
alimentos_df['subject_num'] = alimentos_df['subject'].str.extract(r'(\d+)').astype(int)

# Ordenar stimname según tu lista original
alimentos_df['stimname'] = pd.Categorical(
    alimentos_df['stimname'],
    categories=stimname_variables,
    ordered=True
)

# Orden final
alimentos_df = alimentos_df.sort_values(by=['subject_num', 'stimname']).reset_index(drop=True)

print(alimentos_df.head(20))
```

```python
# archivo final ALIMENTOS, activación del eeg
import numpy as np

# 1) Cargar todos los .set con glob
set_files = glob.glob(
    r"C:\Users\marub\OneDrive\Escritorio\Datos proyecto modelos I\EEG\*.set"
)

eeg_dict = {}  # subject → raw

for file_path in set_files:
    filename = os.path.basename(file_path)
    match = re.search(r"(sub-\d+)", filename)
    subject_id = match.group(1)
    
    print(f"Cargando EEG de {subject_id}...")
    raw = mne.io.read_raw_eeglab(file_path, preload=True)
    eeg_dict[subject_id] = raw


# 2) Extraer activación por evento
rows = []

for _, ev in alimentos_df.iterrows():
    subject = ev["subject"]
    t = ev["time_stimon"]

    if subject not in eeg_dict:
        print(f"⚠️ Advertencia: No hay EEG para {subject}")
        continue

    raw = eeg_dict[subject]
    
    data = raw.get_data()     # canales × tiempo
    times = raw.times         # vector de tiempos
    ch_names = raw.ch_names

    # índice más cercano al tiempo del evento
    idx = np.argmin(np.abs(times - t))
    
    # señal en ese instante (todos los canales)
    signal_at_t = data[:, idx]

    # armar fila final
    row = {
        "subject": subject,
        "stimname": ev["stimname"],
        "time_stimon": float(t),
        "time_eeg": float(times[idx])
    }

    for ch_name, value in zip(ch_names, signal_at_t):
        row[ch_name] = value

    rows.append(row)


# 3) DataFrame final
alimentos_eeg = pd.DataFrame(rows)

print("Activaciones extraídas:", alimentos_eeg.shape)
print(alimentos_eeg.head())
```
