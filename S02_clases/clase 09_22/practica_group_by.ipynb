```python
tipos_sujetos = ['S' + str(i).zfill(2) for i in range (4)]
generos = ['M', 'F']
condicion = ['control', 'experimental']
region = ['hipocampo', 'cortex prefrontal', 'amigdala']
neurotransmisores =  ['dopamina', 'serotonina']

#Crear un data framde de 200 filas
#con las columnas: tipos_sujetos, generos, confición, región, neurotransmisor;
#estas columnas deben tener una distribución normal
#agregar una columna llamada "precisón" q tenga una ditribución normal con mu = 0.92 y sigma = 0.08
#Este dataset pretende simular los resultados de un aestidio en el que se miden variables cognitivas y fisiológicas en
#diferentes regiones cerebrels, sujetos y condicones experimentales
```

```python
df_neuro = pd.DataFrame(columns = ['tipos_sujetos', 'generos', 'condicion', 'region', 'neurotransmisores'], index = range(200))
```

```python
df_neuro
```

```python
longitud = 200
entero = np.random.randint(0,4, size = longitud)
sujetos = ['S' + str(i).zfill(2) for i in entero]
```

```python
df_neuro['tipos_sujetos'] = sujetos
```

```python
df_neuro
```

```python
ops_generos = np.random.randint(len(generos), size = longitud)
generos2 = [generos[i] for i in ops_generos]
generos2
df_neuro['generos'] = generos2
df_neuro
```

```python
ops_condicions = np.random.randint(len(condicion), size = longitud)
condicion2 = [condicion[i] for i in ops_condicions]
df_neuro['condicion'] = condicion2
df_neuro
```

```python
ops_regiones = np.random.randint(len(region), size = longitud)
region2 = [region[i] for i in ops_regiones]
df_neuro['region'] = region2
df_neuro
```

```python
ops_neuro = np.random.randint(len(neurotransmisores), size = longitud)
neuro2 = [neurotransmisores[i] for i in ops_neuro]
df_neuro['neurotransmisores'] = neuro2
df_neuro
```

```python
df_neuro['precisión'] = np.random.normal(0.92,0.08, size = 200)
```
