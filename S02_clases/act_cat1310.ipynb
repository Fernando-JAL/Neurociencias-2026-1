```python
file_path = r"C:\Users\HP\Documents\GitHub\Neurociencias-2026-1\S03_datasets\BCT\cat.mat"

file_dict = scipy.io.loadmat(file_path)
### para leer el archivo en .mat
```

```python
file_dict
```

```python
matriz = file_dict ['CIJctx']
columnas = file_dict ['Names']
```

```python
cat_df = pd.DataFrame(data=matriz, columns=columnas, index=columnas)
```

```python
plt.figure(figsize=(4,4))
ax = sns.heatmap(cat_df, 
                 annot=False, cmap='hot', fmt='.2f',
                 xticklabels=cat_df.columns, yticklabels=cat_df.columns)
ax.set(xlabel='canales', ylabel='canales')
plt.show()
```

```python
aplanada = np.ravel(cat_df) 
hist = plt.hist(aplanada, bins=4)
hist
```

```python
valores, conteos = np.unique(aplanada, return_counts=True)
```

```python
porcentajes = (conteos / np.sum(conteos)) * 100
porcentajes
```
