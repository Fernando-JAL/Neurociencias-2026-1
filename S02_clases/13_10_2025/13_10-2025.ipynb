```python
matriz= file_dict['CIJctx']
columnas = file_dict['Names']
```

```python
matriz.shape, len(columnas) 
```

```python
gato_df = pd.DataFrame(data=matriz, columns=columnas, index=columnas)
```

```python
gato_df.columns = [columna.split(' ')[0] for columna in gato_df.columns]
gato_df.index = gato_df.columns
```

```python
ax = sns.heatmap(
    gato_df,
    annot=False,       
    fmt='0f',          
    cmap='hot',
    xticklabels=gato_df.columns,
    yticklabels=gato_df.columns
)
plt.show()
```

```python
print("Tama√±o del DataFrame:", gato_df.shape)
print("Primeros nombres de canales:", gato_df.columns[:5].tolist())

# Generar mapa de calor con los nombres
plt.figure(figsize=(10,8))
sns.heatmap(
    gato_df,
    annot=False,
    cmap='hot',
    xticklabels=gato_df.columns,
    yticklabels=gato_df.index
)
plt.title("Mapa de calor de CIJctx con nombres de canales")
plt.xlabel("Canales")
plt.ylabel("Canales")
plt.show()
```
