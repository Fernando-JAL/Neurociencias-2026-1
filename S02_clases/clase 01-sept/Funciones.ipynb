## Funciones recursivas

```python
def cuenta_atras (n):
    if n>0: ##tiene que ver algo que evite que se mande a llamar de forma infinita
        print (n)
        cuenta_atras(n-1)
    else: 
        print("booooom!!")
    print ("fin de la función", n)

cuenta_atras(3)
```

## hacer en recursiva 
factorial 
n! = n * (n-1)!
