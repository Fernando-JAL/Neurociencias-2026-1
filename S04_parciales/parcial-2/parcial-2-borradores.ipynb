```python
# 1. Visualizar .mat en Pyhton

# Requiere: scipy
from scipy.io import loadmat

# Cargar archivo .mat
mat_data = loadmat('Coactivation_matrix.mat')

# Mostrar todas las variables y su contenido
for key, value in mat_data.items():
    if not key.startswith('__'):  # omite metadatos internos
        print(f"\nðŸ”¹ Variable: {key}")
        print(f"Tipo: {type(value)}")
        print(f"Forma: {getattr(value, 'shape', 'N/A')}")
        print("Contenido:")
        print(value)
```

```python
# 2. Convertir .mat a .xlsx

# Requiere: scipy (ya usado arriba) y pandas
import pandas as pd

mat_data = loadmat('Coactivation_matrix.mat')

# Exportar cada variable a una hoja de Excel
with pd.ExcelWriter('Coactivation_matrix.xlsx') as writer:
    for key, value in mat_data.items():
        if not key.startswith('__'):
            try:
                df = pd.DataFrame(value)
                df.to_excel(writer, sheet_name=key, index=False)
            except Exception as e:
                print(f"No se pudo exportar '{key}': {e}")
```

```python
# 3. VIsualizar .edf en Python

# Requiere: mne
import mne

# Cargar archivo EDF
raw = mne.io.read_raw_edf('chb01_01.edf', preload=True)

# Mostrar informaciÃ³n general
print(raw.info)
print("Canales:", raw.ch_names)
print(f"DuraciÃ³n: {raw.times[-1]:.2f} segundos")
print(f"Frecuencia de muestreo: {raw.info['sfreq']} Hz")
```

```python
# 4. Convertir .edf a .xlsx

# Requiere: pandas y mne (ya usados arriba)

# Extraer datos y tiempos
data, times = raw.get_data(return_times=True)

# Crear DataFrame sin modificar datos
df = pd.DataFrame(data.T, columns=raw.ch_names)
df.insert(0, 'Time (s)', times)

# Exportar a Excel
df.to_excel('chb01_01.xlsx', index=False)
```
