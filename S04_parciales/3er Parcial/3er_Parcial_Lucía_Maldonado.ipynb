```python
# Importar las librer√≠as
import numpy as np
import pandas as pd
import networkx as nx
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.io
import os
import mne
from scipy.signal import coherence, hilbert
```

```python
# Abrir la carpeta
folder_path= r"C:\Users\jumma\OneDrive\Documentos\GitHub\Neurociencias-2026-1\S03_datasets\eeg-motor-movementinmagery"
# Obtener paths de la carpeta
def lista_paths(folder_path):
    lista_de_paths= []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".xlsx"):
                lista_de_paths.append(os.path.join(root, file))
    return lista_de_paths
# leer cada archivo de la carpeta
def leer_todos_excel(lista_de_paths):
    ensayos = {}
    for excel_path in lista_de_paths:
        ensayo = os.path.basename(excel_path)
        ensayos[ensayo] = pd.read_excel(excel_path, sheet_name=None, index_col=0)
    return ensayos
```

```python
paths= lista_paths(folder_path)
```

```python
diccionario_ensayos= leer_todos_excel(paths)
```

```python
# Leer el archivo de las coordenadas
eeg_coords_csv = r"C:\Users\jumma\OneDrive\Documentos\GitHub\Neurociencias-2026-1\S03_datasets\MI_coordinates.csv"
eeg_coords = pd.read_csv(eeg_coords_csv)
eeg_coords.set_index("canal", drop=True, inplace=True)
```

```python
#obtener los dataframes
def leer_df_ensayo(lista_de_paths):
    ensayos = {} 
    for excel_path in lista_de_paths:
        hojas = pd.read_excel(excel_path, sheet_name=None, index_col=0)
        for nombre_ensayo, df in hojas.items():
            if nombre_ensayo not in ensayos:
                ensayos[nombre_ensayo] = []  
            ensayos[nombre_ensayo].append(df) 
    return ensayos
```

```python
df_ensayos= leer_df_ensayo(paths)
```

```python
#stacks de cada ensayo
def hacer_stacks(ensayos):
    stacks = {}
    for ensayo, lista_dfs in ensayos.items():
        stack = np.stack([df.values for df in lista_dfs])
        stacks[ensayo] = stack
    return stacks
```

```python
diccionario_de_stacks= hacer_stacks(df_ensayos)
```
