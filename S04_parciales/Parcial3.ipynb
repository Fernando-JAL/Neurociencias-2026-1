```python
import numpy as np
import pandas as pd
import networkx as nx

import seaborn as sns
import matplotlib.pyplot as plt
import scipy.io as sp
from scipy.io import loadmat
import mne
import glob
import os

# Ruta donde están TODOS los .mat
ruta = r"C:\Datasets\eeg\Proyecto\Preprocessed\DecisionMaking\*.mat"

# Diccionario final donde guardaremos cada archivo
mat_files = {}

# Recorremos todos los .mat encontrados con glob
for archivo in glob.glob(ruta):
    nombre = os.path.basename(archivo)       # ejemplo: "Player_sub01_Observer_sub02.mat"
    contenido = loadmat(archivo)             # se convierte en diccionario
    mat_files[nombre] = contenido            # guardamos en el diccionario principal

# Mostrar llaves cargadas
mat_files.keys()
```
