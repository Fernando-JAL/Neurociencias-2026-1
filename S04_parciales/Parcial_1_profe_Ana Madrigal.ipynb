üîπ Ejercicio Frecuencia de n√∫meros en matriz

Genera una matriz 6x6 de enteros aleatorios entre 1 y 10 usando NumPy.

Recorre la matriz y guarda en un diccionario la frecuencia de cada n√∫mero.

Convierte el diccionario en una lista de tuplas (n√∫mero, frecuencia) y ord√©nala de mayor a menor frecuencia.

Imprime la matriz, el diccionario y la lista ordenada.

```python
## matriz 6x6 
import numpy as np
matriz=np.random.randint(1,11,size=[6,6])
print(matriz)

########
frecuencias={}
for fila in matriz:
    for numero in fila:
        num_normal = int(numero) 
        if num_normal in frecuencias:
            frecuencias[num_normal] += 1
        else:
            frecuencias[num_normal] = 1
print(frecuencias)
lista= list(frecuencias.items())

lista_primero_frec = []
for numero in frecuencias:
    veces = frecuencias[numero]
    lista_primero_frec.append((veces, numero))
numeros_ordenados = sorted(lista_primero_frec, reverse=True)
print(numeros_ordenados) 

```

üîπ Ejercicio Matriz de distancias

Genera una matriz 10x2 con coordenadas aleatorias (x, y) entre 0 y 100.

Con un bucle, calcula las distancias euclidianas entre cada par de puntos y gu√°rdalas en un diccionario con claves (i,j).

Convierte el diccionario en un DataFrame de 10x10 (matriz de distancias).

Encuentra el par de puntos m√°s cercanos y el m√°s lejano.

```python
import numpy as np
matriz= np.random.randint(0, 101, size=(10, 2))
matriz
distancias= {}

for i in range(10):
    for j in range(i+1, 10):
        x1, y1 = matriz[i]
        x2, y2 = matriz[j]
        dist = ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5
        distancias[(i, j)] = dist

print(distancias)

import pandas as pd
# Convertir a DataFrame
filas = []
for (i, j), d in distancias.items():
    filas.append([i, j, d])

df = pd.DataFrame(filas, columns=["Punto i", "Punto j", "Distancias"])
print(df)
# Distancia m√≠nima y m√°xima
dist_min = df["Distancias"].min()
dist_max = df["Distancias"].max()

print(f"Distancia maxima={dist_max},Distancia minima={dist_min}")
```

Guarda las medias en un diccionario {Mes: Promedio}.

Convierte el diccionario en una Series y ord√©nala de mayor a menor.

```python
import numpy as np
diaria=np.random.randint(10,41,size=(365))
print(diaria)
import pandas as pd
serie_diaria=pd.Series(diaria,index=range(1,len(diaria)+1))
print(serie_diaria)

##Data frame
data_final=serie_diaria.reset_index()
data_final.columns = ["D√≠a", "Temperatura"]
data_final
```

```python
##Medias cada mes 
###########################################
dias_mes=30
medias={}
for mes in range(12):
    inicio= mes*dias_mes
    fin=(mes+1)*dias_mes
    medias[f"Mes{mes+1}"]= data_final["Temperatura"][inicio:fin].mean()

print(medias)
##Convierte el diccionario en una Series y ord√©nala de mayor a menor.
import pandas as pd
serie=pd.Series(medias).sort_values(ascending=False)
print(serie)


```

üîπ Ejercicio EEG por canales

Simula una matriz 500x8 con NumPy representando 500 muestras de se√±al EEG en 8 canales (valores flotantes entre -100 y 100 ¬µV).

Convierte la matriz en un DataFrame con columnas "Canal_1"... "Canal_8".

Usa un bucle para calcular la potencia media de cada canal (media del cuadrado de la se√±al).

Guarda los resultados en un diccionario y convi√©rtelo en una Series ordenada.

Identifica el canal con mayor potencia.

```python
import numpy as np
EEG=np.random.rand(500,8)*200-100
print(EEG)

import pandas as pd
matriz_EEG=pd.DataFrame(EEG,columns=["Canal_1","Canal_2","Canal_3","Canal_4","Canal_5","Canal_6","Canal_7","Canal_8"])
print(matriz_EEG)

potencias_medias={}
for canal in matriz_EEG.columns:
    resultado=matriz_EEG[canal].values
    medias_cuadrado=np.mean(resultado**2)
    potencias_medias[canal]=medias_cuadrado
print(potencias_medias)

serieEEG=pd.Series(potencias_medias)
print(serieEEG)
canal_mayor_frecuencia=serieEEG.idxmax()
valor_mayor_frecuencia = serieEEG.max()
print(f"{canal_mayor_frecuencia}:{valor_mayor_frecuencia}")
```

üîπ Ejercicio An√°lisis de respuestas a est√≠mulos

Genera un DataFrame con 30 sujetos (filas) y 3 condiciones experimentales ("Cond_A", "Cond_B", "Cond_C") con valores de tiempo de reacci√≥n simulados (valores entre 200‚Äì800 ms).

Usa un bucle para calcular el promedio por condici√≥n en toda la muestra.

Convierte los resultados en un diccionario y luego en una Series.

Agrega una columna "Mejor_condici√≥n" al DataFrame indicando la condici√≥n m√°s r√°pida por sujeto.

Cuenta cu√°ntos sujetos respondieron mejor en cada condici√≥n.

```python
import numpy as np
respuestas_estimulos=np.random.randint(200,801, size=(30,3))
print(respuestas_estimulos)
import pandas as pd
data= pd.DataFrame(respuestas_estimulos, index=range(1,len(respuestas_estimulos)+1),columns=["Cond_A","Cond_B","Cond_C"])
print(data)

resultados_promedio={}
for valor in data.columns:
    resultados=data[valor].values
    promedio=np.mean(resultados)
    resultados_promedio[valor]=promedio
print(resultados_promedio)

Serie_promedio= pd.Series(resultados_promedio)
print(Serie_promedio)

###################### 
# agregar columna data.frame
maxdelomax=[]
for valor in data.index:
    resultado2=data.loc[valor]
    maxdelomax.append(resultado2.idxmax())
    print(f"{maxdelomax}")
data["Mejor_condici√≥n"]=maxdelomax
print(data)

##cuantos sujetos responderion mejor en que condici√≥n
A=maxdelomax.count("Cond_A")
B=maxdelomax.count("Cond_B")
C=maxdelomax.count("Cond_C")
print(f"{A} sujetos respondieron mejor a la condici√≥n A, {B} sujetos respondieron mejor a la condici√≥n B, {C} sujetos respondieron mejor a la condici√≥n C")
```
