## Ejercicio Frecuencia de números en matriz

```python
##Genera una matriz 6x6 de enteros aleatorios entre 1 y 10 usando NumPy.
import numpy as np
matriz= np.random.uniform(1,11, size=[6,6])
matriz_enteros =(matriz.astype(int))
matriz_enteros
```

```python
##Recorre la matriz y guarda en un diccionario la frecuencia de cada número.
frecuencias ={}
for fila in matriz_enteros:
    for valor in fila:
        if valor in frecuencias:
            frecuencias[valor] += 1
        else:
            frecuencias[valor] = 1


print(frecuencias)
```

```python
#Convierte el diccionario en una lista de tuplas (número, frecuencia) y ordénala de mayor a menor frecuencia.
lista= ([7,10], [8,5], [10,4], [3,4], [1,3], [5,2], [6,3], [9,2], [4,3])
sorted(lista)
lista_ordenada = sorted(lista, key=lambda x: x[1], reverse=True)
lista_ordenada
```

```python
#Imprime la matriz, el diccionario y la lista ordenada.
print(f" matriz {matriz_enteros}")
print(f" Diccionario frecuencias:   {frecuencias}")
print(f" Frecuencias ordenadas {lista_ordenada}")
```

## Ejercicio Clasificación de temperaturas

```python
#Genera una Serie con 50 valores aleatorios entre -10 y 40 (temperaturas).
import pandas as pd
import numpy as np
serie= (pd.Series(np.random.uniform( -10, 40, size=[50])))
serie
```

```python
#Recorre la Serie y clasifica cada temperatura en un diccionario:
#"frío" si < 1, "templado" si 10–25, "calor" si > 25
resultado = {"Temperatura": [], "Categoría": []}

for temp in serie:
    resultado["Temperatura"].append(temp)
    
    if temp < 10:
        resultado["Categoría"].append("frío")
    elif temp <= 25: 
        resultado["Categoría"].append("templado")
    else:
        resultado["Categoría"].append("calor")
resultado     
```

```python
# Convierte el diccionario en un DataFrame con columnas "Temperatura" y "Categoría".
df=pd.DataFrame(resultado)
df
```

```python
#Cuenta cuántos valores cayeron en cada categoría.
conteo = df["Categoría"].value_counts()
print(conteo)
```

## Ejercicio Matriz de distancias

```python
#Genera una matriz 10x2 con coordenadas aleatorias (x, y) entre 0 y 100.
matriz=np.random. uniform(0,100, size=[10,2])
matriz
```

```python
## Con un bucle, calcula las distancias euclidianas entre cada par de puntos y guárdalas en un diccionario con claves (i,j)
distancias= {}
for i in range (10):
    for j in range (i+1, 10):
        x1, y1 = matriz[i]
        x2, y2 = matriz[j]
        
        d = ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5  
        
        distancias[(i,j)] = d   
```

```python
#Convierte el diccionario en un DataFrame de 10x10 (matriz de distancias).
import pandas as pd
filas=[]
for (i,j), d in distancias.items():
    filas.append([i, j, d])
df= pd.DataFrame (filas, columns=("Punto i", "Punto j ", "Distancias"))
df
```

```python
#Encuentra el par de puntos más cercanos y el más lejano.
dist_min = df["Distancias"].min()
dist_max = df["Distancias"].max()

print(f"los puntos más cercano es 1, 8, y su distancia es {dist_min}")
print (f"Los puntos más lejanos son  0,3 y su distancia es {dist_max}")
```

## Ejercicio Análisis de series temporales

```python
#Genera con NumPy una Serie de 365 valores aleatorios entre 10 y 40 (temperatura diaria de un año).
temp=np.random.uniform(10,40, size=[365])
días= np.arange(365)
```

```python
#Convierte en un DataFrame con columnas "Día" y "Temperaturas"
data=pd.DataFrame({"días del año":días,"temperatura": temp})
```

```python
medias = []

for mes in range(12):  
    inicio = mes * 30
    fin = inicio + 30
    temperaturas_mes = temp[inicio:fin] 
    media_mes = sum(temperaturas_mes) / 30
    medias.append(media_mes)
medias
```

```python
#Guarda las medias en un diccionario {Mes: Promedio}.
diccionario={}
meses=["enero", "febrero", "marzo", "abril", "mayo", "junio", "julio", "agosto", "septiempre", "octubre", "noviembre", "diciembre"]
for i in range (len(medias)):
     clave = meses[i]
     diccionario[clave]= medias[i]
diccionario
```

```python
#Convierte el diccionario en una Series y ordénala de mayor a menor.
serie=pd.Series(diccionario)
serie.sort_values(ascending=False)
```

## Ejercicio Conectividad funcional (correlaciones)

```python
#Con un bucle, calcula la correlación de Pearson entre todas las parejas de áreas 
#y guárdala en un diccionario con claves (Área_i, Área_j).

diccionario={}
for area_i in dt:
    for area_j in dt:
        diccionario[(area_i, area_j)] =  dt[area_i].corr(dt[area_j])
diccionario
```

```python
#Convierte el diccionario en un DataFrame cuadrado (matriz de conectividad).
matriz = dt.corr()
matriz
```

##  Ejercicio EEG por canales.

```python
#Simula una matriz 500x8 con NumPy representando 500 muestras de señal EEG en
#8 canales (valores flotantes entre -100 y 100 µV).
matriz=np.random.uniform(-100, 100, (500,8))
#Convierte la matriz en un DataFrame con columnas "Canal_1"... "Canal_8".
df=pd.DataFrame(matriz, columns=("Canal 1", "Canal 2", "Canal 3", "Canal 4", "Canal 5", "Canal 6", "Canal 7", "Canal 8"))
df
```

```python
#Usa un bucle para calcular la potencia media de cada canal (media del cuadrado de la señal).
potencia= {}
for columna in df.columns:                            
        potencia[columna]= (df[columna]**2).mean()
potencia
#Guarda los resultados en un diccionario y conviértelo en una Series ordenada.
serie_pot=pd.Series(potencia)
serie_pot
```

```python
#Identifica el canal con mayor potencia. 
mayor_pot= serie_pot.max()
print(f"El canal con mayor potencia es el canal 1, con {mayor_pot} µV")
```

# Ejercicio Análisis de respuestas a estímulos

```python
# Genera un DataFrame con 30 sujetos (filas) y 3 condiciones experimentales ("Cond_A", "Cond_B", "Cond_C")
# con valores de tiempo de reacción simulados (valores entre 200–800 ms).
sujetos=np.random.uniform(200, 800, (30,3))
estímulos=pd.DataFrame(sujetos, columns=["cond_A","Cond_B", "Cond_C"])
```

```python
#Usa un bucle para calcular el promedio por condición en toda la muestra.
media={}
for condición in estímulos.columns:
     media[condición]= (estímulos[condición]).mean()
media
#Convierte los resultados en un diccionario y luego en una Series
serie=pd.Series(data=media)
serie
```

```python
#Agrega una columna "Mejor_condición" al DataFrame indicando la condición más rápida por sujeto.
lista=[]
lista2=[]
for i, fila in estímulos.iterrows():
    lista.append(fila.max())
    lista2.append(fila.idxmax())
lista   
estímulos['Mejor condición']= lista
estímulos["Origen"]=  lista2
```

```python
#Cuenta cuántos sujetos respondieron mejor en cada condición.
estímulos["Origen"].value_counts()
```
