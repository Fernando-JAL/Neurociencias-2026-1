```python
import mne
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
from scipy.signal import coherence, hilbert
from scipy.io import loadmat
import community.community_louvain as community_louvain
from matplotlib import cm
```

### 1. Utilizando el dataset de Coactivation_matrix.mat:

```python
coactivacion =  loadmat (r"C:\Users\HP\Documents\GitHub\Neurociencias-2026-1\S03_datasets\BCT\Coactivation_matrix.mat")
```

```python
coactivacion.keys()
```

```python
matriz = coactivacion['Coactivation_matrix']
matriz.shape
```

```python
## visualizar los valores minimos y maximos de la matriz para determinar el umbral
print(type(matriz))
print(matriz.dtype)

print("Valores únicos:", np.unique(matriz))
print("Máximo:", np.max(matriz))
print("Mínimo:", np.min(matriz))
```

```python
## la umbralice por densidad
def threshold_by_density(matrix, density=0.1):
    flat = matrix.flatten()
    cutoff = np.percentile(flat, 100 * (1 - density))
    return (matrix > cutoff).astype(int)

binary_matrix = threshold_by_density(matriz, density=0.1)
binary_matrix
```

```python
G = nx.from_numpy_array(binary_matrix)

C = nx.average_clustering(G)
L = nx.average_shortest_path_length(G)

G_rand = nx.gnm_random_graph(G.number_of_nodes(), G.number_of_edges())

C_rand = nx.average_clustering(G_rand)
L_rand = nx.average_shortest_path_length(G_rand)

# coeficiente de mundo pequeño
sigma = (C / C_rand) / (L / L_rand)
print(f"Coeficiente de mundo pequeño: {sigma:.3f}")

# indica propiedades de mundo pequeño
```

```python
# Calcule las comunidades del grafo
partition = community_louvain.best_partition(G)

num_communities = len(set(partition.values()))
print(f"Número de comunidades detectadas: {num_communities}")
```

```python
pos = nx.spring_layout(G)
colors = [partition [node] for node in G.nodes()]
nx.draw(G, pos, node_color=colors, with_labels=True, cmap=plt.cm.Set3)
plt.title("Comunidades detectadas (Louvain)")
plt.show()
```

```python
# Eficiencia global
global_eff = nx.global_efficiency(G)
print(f"Eficiencia global: {global_eff:.4f}")

# Eficiencia local
local_eff = nx.local_efficiency(G)
print(f"Eficiencia local: {local_eff:.4f}")
```

```python
# Determine el grado de cada nodo
grados = dict(G.degree())

for nodo, grado in grados.items():
    print(f"Nodo {nodo}: grado {grado}")
```

```python
# Calcule los hub
hubs_por_grado = sorted(grados.items(), key=lambda x: x[1], reverse=True)
print("Top 5 hubs por grado:")
for nodo, grado in hubs_por_grado[:5]:
    print(f"Nodo {nodo}: grado {grado}")
```

```python
# Cree los 3 grafos 2D filtrando la matriz de conectividad con los valores dados por los percentiles del ejercicio anterior
df = pd.DataFrame(matriz)

f_p25 = df.copy()
f_p25[f_p25 < p25] = 0
f_p25[f_p25 >= p25] = 1

f_p50 = df.copy()
f_p50[f_p25 < p50] = 0
f_p50[f_p25 >= p50] = 1

f_p75 = df.copy()
f_p75[f_p25 < p75] = 0
f_p75[f_p25 >= p75] = 1
```

3. Utilizando el dataset de Coactivation_matrix.mat:
* Genere el grafo 3D
* Haga que el tamaño de nodos sea proporcional a su grado; es decir, q entre mayor sea su grado, mayor sea el tamaño del nodo ploteado
* Haga que el color de las aristas este relacionado al valor de la matriz de conectividad. Utilizar el map color Hot

```python
matriz = coactivacion['Coactivation_matrix']
coorde = coactivacion['Coord']
```

```python
coordenadas_df = pd.DataFrame(coorde, columns=['x', 'y', 'z'])
coordenadas_df.index = range(len(coordenadas_df))
G = nx.from_numpy_array(matriz)
```

```python
pos3D= {canal: (coordenadas_df.loc[canal, 'x'],
                coordenadas_df.loc[canal, 'y'],
                coordenadas_df.loc[canal, 'z']) for canal in coordenadas_df.index}

fig = plt.figure(figsize=(4,4))
## definir el tamaño
ax = fig.add_subplot(111, projection='3d')
## contar los subplots que voy a tener y que de esos vaya uno por uno (111)

## dibujar nodos en 3d
for canal, (x, y, z) in pos3D.items():
    ax.scatter(x, y, z, s=100)
    ax.text(x, y, z+0.03, canal, fontsize=10, ha='center')

for i, j, data in G.edges(data=True):
    x= [pos3D[i][0], pos3D[j][0]]
    y= [pos3D[i][1], pos3D[j][1]]
    z= [pos3D[i][2], pos3D[j][2]]
    ax.plot(x, y, z, linewidth=data['weight']*5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Red de conectividad (3D)')
plt.show()
```

```python
# Haga que el tamaño de nodos sea proporcional a su grado; es decir, q entre mayor sea su 
#grado, mayor sea el tamaño del nodo ploteado
grados = dict(G.degree(weight='weight')) 
grados_array = np.array(list(grados.values()))

min_size = 50
max_size = 500

sizes = min_size + (grados_array - grados_array.min()) / (grados_array.max() - grados_array.min()) * (max_size - min_size)

pos3D = {
    canal: (coordenadas_df.loc[canal, 'x'],
            coordenadas_df.loc[canal, 'y'],
            coordenadas_df.loc[canal, 'z'])
    for canal in coordenadas_df.index
}

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

for i, (canal, (x, y, z)) in enumerate(pos3D.items()):
    ax.scatter(x, y, z, s=sizes[i], color='skyblue', alpha=0.8, edgecolor='k')
    ax.text(x, y, z+0.03, str(canal), fontsize=8, ha='center')

for i, j, data_edge in G.edges(data=True):
    x = [pos3D[i][0], pos3D[j][0]]
    y = [pos3D[i][1], pos3D[j][1]]
    z = [pos3D[i][2], pos3D[j][2]]
    w = data_edge.get('weight', 0.1)
    ax.plot(x, y, z, color='gray', linewidth=w * 3, alpha=0.7) 

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Red de conectividad cerebral (nodos proporcionales al grado)')
plt.show()
```

```python
# Haga que el color de las aristas este relacionado al valor de la matriz de conectividad. 
#Utilizar el map color Hot
grados = dict(G.degree(weight='weight'))
grados_array = np.array(list(grados.values()))
min_size, max_size = 20, 100
sizes = min_size + (grados_array - grados_array.min()) / (grados_array.max() - grados_array.min()) * (max_size - min_size)

pos3D= {
    canal: (coordenadas_df.loc[canal, 'x'],
            coordenadas_df.loc[canal, 'y'],
            coordenadas_df.loc[canal, 'z'])
    for canal in coordenadas_df.index
}

pesos = np.array([d['weight'] for (_, _, d) in G.edges(data=True)])
norm = plt.Normalize(vmin=pesos.min(), vmax=pesos.max())
cmap = plt.colormaps['hot'] 

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

for i, (canal, (x, y, z)) in enumerate(pos3D.items()):
    ax.scatter(x, y, z, s=sizes[i], color='skyblue', alpha=0.9, edgecolor='k')
    ax.text(x, y, z+0.03, str(canal), fontsize=8, ha='center')

for i, j, data_edge in G.edges(data=True):
    w = data_edge['weight']
    color = cmap(norm(w))  
    x = [pos3D[i][0], pos3D[j][0]]
    y = [pos3D[i][1], pos3D[j][1]]
    z = [pos3D[i][2], pos3D[j][2]]
    ax.plot(x, y, z, color=color, linewidth=2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Red de conectividad cerebral (3D) color')

sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
fig.colorbar(sm, ax=ax, shrink=0.6, label='Peso de conexión')

plt.show()
```

```python
raw = mne.io.read_raw_edf(chb, preload=True)
raw.resample(sfreq_target)
raw.filter(1., 45., fir_design='firwin')
data = raw.get_data()
ch_names = raw.info['ch_names']
n_channels = len(ch_names)
```
