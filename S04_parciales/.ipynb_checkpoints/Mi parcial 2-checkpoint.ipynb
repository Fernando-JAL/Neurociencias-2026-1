```python
import mne
import numpy as np
import pandas as pd
from scipy.signal import coherence, hilbert
import networkx as nx
import matplotlib. pyplot as plt
import scipy.io 
import seaborn as sns
from scipy.io import loadmat
from matplotlib import cm
```

```python
coa=r"C:\Users\ecere\Documents\GitHub\Neurociencias-2026-1\S03_datasets\BCT\Coactivation_matrix.mat"
coa_dict=scipy.io.loadmat(coa)
```

```python
#Coeficiente mundo pequeño
#Coactiviation_matriz.max
mat=coa_dict['Coactivation_matrix']
mat_df=pd.DataFrame(mat)
G=nx.from_pandas_adjacency( mat_df)
```

```python
C=nx.average_clustering(G)
Gcc= G.subgraph(max(nx.connected_components(G), key=len))
L=nx.average_shortest_path_length(Gcc)
```

```python
Gr = nx.gnm_random_graph(G.number_of_nodes(), G.number_of_edges())
Cr = nx.average_clustering(Gr)
Gcc_r = Gr.subgraph(max(nx.connected_components(Gr), key=len))
Lr = nx.average_shortest_path_length(Gcc_r)
```

```python
sigma = (C/Cr) / (L/Lr)
print(f"σ = {sigma:.3f} → {'Mundo pequeño' if sigma>1 else 'No es mundo pequeño'}")
```

```python
#Comunidades del grafo 
from networkx.algorithms.community import greedy_modularity_communities

comunidades = greedy_modularity_communities(G)
print(f"Comunidades encontradas: {len(comunidades)}")
for i, c in enumerate(comunidades):
    print(f"Comunidad {i+1}: {list(c)}")
```

```python
#Hubs
grados = dict(G.degree())
umbral= np.percentile(list(grados.values()),90)
hubs=[n for n, g in grados.items() if g >= umbral]
hubs
```

```python
#Eficiencia global 
def efi_global(G):
    inv_dist = [1/d for comp in nx.connected_components(G)
                for s, t_d in nx.all_pairs_shortest_path_length(G.subgraph(comp))
                for d in t_d.values() if d>0]
    return np.mean(inv_dist)
Eg= efi_global(G)
```

```python
Eg
```

```python
#Eficiencia local
def efi_local(G):
    return np.mean([efi_global(G.subgraph(G.neighbors(n)))
                    for n in G if len(list(G.neighbors(n)))>1])
El= efi_local(G)
```

```python
#  dataset de chb01_01.edf:
chb01=r"C:\Users\ecere\Documents\GitHub\Neurociencias-2026-1\S03_datasets\EEG_practica\chb01_01.edf"
sfreq_target = 256           
fmin, fmax = 8, 13 
raw = mne.io.read_raw_edf(chb01, preload=True)
raw.resample(sfreq_target)
raw.filter(1., 45., fir_design='firwin')
data = raw.get_data()
ch_names = raw.info['ch_names']
n_channels = len(ch_names)
```

```python
#Mundo pequeño 
corr_mat = np.corrcoef(data)
threshold = np.percentile(corr_mat, 50)
adj_matrix = np.where(corr_mat >= threshold, 1, 0)
G = nx.from_numpy_array(adj_matrix)
C = nx.average_clustering(G)
try:
    L = nx.average_shortest_path_length(G)
except nx.NetworkXError:
    # si el grafo no es conexo, tomar camino promedio de la componente gigante
    largest_cc = max(nx.connected_components(G), key=len)
    G_lcc = G.subgraph(largest_cc)
    L = nx.average_shortest_path_length(G_lcc)
    G_rand = nx.gnm_random_graph(n_channels, G.number_of_edges())
C_rand = nx.average_clustering(G_rand)
try:
    L_rand = nx.average_shortest_path_length(G_rand)
except nx.NetworkXError:
    largest_cc_rand = max(nx.connected_components(G_rand), key=len)
    G_lcc_rand = G_rand.subgraph(largest_cc_rand)
    L_rand = nx.average_shortest_path_length(G_lcc_rand)
    sigma = (C / C_rand) / (L / L_rand)

print(f"Coeficiente de clustering C: {C:.4f}")
print(f"Camino promedio L: {L:.4f}")
print(f"C_aleatorio: {C_rand:.4f}, L_aleatorio: {L_rand:.4f}")
print(f"mundo pequeño σ: {sigma:.4f}")
```

```python
#Calcule las comunidades del grafo
from networkx.algorithms.community import greedy_modularity_communities

comunidades = greedy_modularity_communities(G)
print(f"Comunidades encontradas: {len(comunidades)}")
for i, c in enumerate(comunidades):
    print(f"Comunidad {i+1}: {list(c)}")
```
