游댳 Ejercicio Frecuencia de n칰meros en matriz

```python
import numpy as np
```

```python
# Genera una matriz 6x6 de enteros aleatorios entre 1 y 10 usando NumPy.
matriz = np.random.randint(1, 11, size=(6, 6))
```

```python
matriz
```

```python
#Recorre la matriz y guarda en un diccionario la frecuencia de cada n칰mero.
```

```python
diccionario = {}
for num in matriz.flatten():  
    diccionario[num] = diccionario.get(num, 0) + 1
```

```python
diccionario
```

```python
#Convierte el diccionario en una lista de tuplas (n칰mero, frecuencia) y ord칠nala de mayor a menor frecuencia.
```

```python
lista = sorted(diccionario.items(), key=lambda x: x[1], reverse=True)
```

```python
lista
```

游댳 Ejercicio Clasificaci칩n de temperaturas

```python
import numpy as np
import pandas as pd
```

```python
# Genera una Serie con 50 valores aleatorios entre -10 y 40 (temperaturas).
```

```python
temperaturas = pd.Series(np.random.randint(-10, 41, size=50))
```

```python
temperaturas 
```

```python
# Recorre la Serie y clasifica cada temperatura en un diccionario:

#"fr칤o" si < 10

#"templado" si 1025

#"calor" si > 25
```

```python
clasificacion_temperaturas = {}
for i, temp in temperaturas.items ():
    if temp < 10:
        clasificacion_temperaturas[i] =(temp, "frio")
    elif 10 <= temp <= 25: 
        clasificacion_temperaturas[i] =(temp, "templado")
    else:
        clasificacion_temperaturas[i] = (temp, "calor") 
```

```python
clasificacion_temperaturas 
```

```python
# Convierte el diccionario en un DataFrame con columnas "Temperatura" y "Categor칤a".
```

```python
df = pd.DataFrame.from_dict(clasificacion_temperaturas, orient="index", columns=["Temperatura", "Categor칤a"])
   
```

```python
df
```

```python
# Cuenta cu치ntos valores cayeron en cada categor칤a.
```

```python
conteo_valores_en_categoria= df["Categor칤a"].value_counts()
```

```python
conteo_valores_en_categoria
```

游댳 Ejercicio Matriz de distancias

```python
import numpy as np
import pandas as pd
```

```python
# Genera una matriz 10x2 con coordenadas aleatorias (x, y) entre 0 y 100.
```

```python
matriz3 = np.random.randint(0, 101, size=(10, 2))
```

```python
matriz3
```

```python
#Con un bucle, calcula las distancias euclidianas entre cada par de puntos y gu치rdalas en un diccionario con claves (i,j).
```

```python
distancias = {}
for i in range(len(matriz3)):
    for j in range(len(matriz3)):
        # distancia euclidiana entre punto i y punto j
        d = np.linalg.norm(matriz3[i] - matriz3[j])
        distancias[(i, j)] = d
```

```python
distancias 
```

```python
# Convierte el diccionario en un DataFrame de 10x10 (matriz de distancias).
```

```python
df = pd.DataFrame(
    np.zeros((10, 10)),
    columns=[f"P{j}" for j in range(10)],
    index=[f"P{i}" for i in range(10)]
)

for (i, j), d in distancias.items():
    df.iloc[i, j] = d
```

```python
df
```

```python
# Encuentra el par de puntos m치s cercanos y el m치s lejano.
```

```python
min_dist = df.replace(0, np.nan).min().min()
max_dist = df.max().max()

pares_min = df.where(df == min_dist).stack().index.tolist()
pares_max = df.where(df == max_dist).stack().index.tolist()
```

```python
pares_min
```

```python
pares_max
```

游댳 Ejercicio An치lisis de series temporales

```python
# Genera con NumPy una Serie de 365 valores aleatorios entre 10 y 40 (temperatura diaria de un a침o).
```

```python
temperatura_a침o = np.random.randint(10, 41, size=365)
```

```python
temperatura_a침o
```

```python
# Convierte en un DataFrame con columnas "D칤a" y "Temperatura"
```

```python
df = pd.DataFrame({
   "D칤a": range(1,366),
"Temperatura" : temperatura_a침o
})
```

```python
df
```

```python
# Usa un bucle para calcular la media de cada mes (asume 30 d칤as por mes).
# Guarda las medias en un diccionario {Mes: Promedio}.
```

```python
medias ={}
for mes in range(1,13):
    inicio = (mes-1)*30
    fin = inicio+30
    promedio = df.loc[inicio:fin-1, "Temperatura"].mean()
    medias[mes] = promedio
```

```python
medias
```

```python
# Convierte el diccionario en una Series y ord칠nala de mayor a menor.
```

```python
serie_medias = pd.Series(medias, name="medias").sort_values(ascending=False)
```

```python
serie_medias
```

游댳 Ejercicio Conectividad funcional (correlaciones)

```python
# Simula con NumPy una matriz 200x4 con series de tiempo de actividad fMRI en 4 치reas cerebrales.
```

```python
act_fMRI= np.random.rand(200, 4)
```

```python
act_fMRI
```

```python
# Convierte en un DataFrame con columnas "츼rea_1"... "츼rea_4".
```

```python
columnas= [f"츼rea_{i}" for i in range (1,5)]
df =pd.DataFrame(act_fMRI, columns=columnas)
```

```python
df
```

```python
# Con un bucle, calcula la correlaci칩n de Pearson entre todas las parejas de 치reas y gu치rdala en un diccionario con claves (츼rea_i, 츼rea_j).
```

```python
correlaciones = {}
for i in range(len(columnas)):
    for j in range(i, len(columnas)):
        area_i = columnas[i]
        area_j = columnas[j]
        corr = df[area_i].corr(df[area_j]) 
        correlaciones[(area_i, area_j)] = corr
        correlaciones[(area_j, area_i)] = corr 
```

```python
correlaciones
```

```python
# Convierte el diccionario en un DataFrame cuadrado (matriz de conectividad).
```

```python
matriz_conectividad = pd.DataFrame(index=columnas, columns=columnas)
for (area_i, area_j), corr in correlaciones.items():
    matriz_conectividad.loc[area_i, area_j] = corr
```

```python
matriz_conectividad 
```
