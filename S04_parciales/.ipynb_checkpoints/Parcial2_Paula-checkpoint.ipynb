```python
import numpy as np
import networkx as nx
import scipy.io
import pandas as pd
```

```python
data_cargar= r"C:\Users\HP\Documents\GitHub\Neurociencias-2026-1\S03_datasets\BCT\Coactivation_matrix.mat.mat"
data=scipy.io.loadmat(data_cargar)
print(data.keys())
matriz = data["Coactivation_matrix"]
df = pd.DataFrame(data=matriz)
df
G = nx.from_pandas_adjacency(df)
```

```python
matriz = np.array(data["Coactivation_matrix"])
matriz = np.squeeze(matriz)

print(matriz.shape)
```

```python
G = nx.from_numpy_array(matriz)
```

```python
##1
##mundo pequeño
def coeficiente_mundo_pequeno(G):
    try:
        C = nx.average_clustering(G)
        L = nx.average_shortest_path_length(G)
        return C / L
    except:
        return None

resultado_mundo_pequeno = coeficiente_mundo_pequeno(G)
if resultado_mundo_pequeno:
    print(round(resultado_mundo_pequeno, 4))
else:
    print("\n1) No se pudo calcular (la red no está conectada)")
```

```python
import networkx.algorithms.community as nx_comm 

def comunidades_grafo(G):
    comunidades = list(nx_comm.greedy_modularity_communities(G))
    return comunidades

comunidades = comunidades_grafo(G)
print("\n2) Comunidades encontradas:", len(comunidades))
for i, c in enumerate(comunidades[:3]):  # solo mostramos algunas
    print("   Comunidad", i + 1, ":", list(c)[:8], "...")
```

```python
def hubs_grafo(G):
    grados = dict(G.degree())
    max_grado = max(grados.values())
    hubs = [n for n, g in grados.items() if g == max_grado]
    return hubs, max_grado

hubs, max_grado = hubs_grafo(G)
print(hubs)
print(max_grado)
```

```python
def eficiencia_global(G):
    n = len(G)
    total = 0
    for i in G:
        for j in G:
            if i != j:
                try:
                    d = nx.shortest_path_length(G, i, j)
                    total += 1 / d
                except:
                    pass
    return total / (n * (n - 1))

E_global = eficiencia_global(G)
print(round(E_global, 4))
```

```python
def eficiencia_local(G):
    valores = []
    for nodo in G:
        vecinos = list(G.neighbors(nodo))
        if len(vecinos) > 1:
            sub = G.subgraph(vecinos)
            valores.append(eficiencia_global(sub))
    if len(valores) > 0:
        return np.mean(valores)
    else:
        return 0

E_local = eficiencia_local(G)
print(round(E_local, 4))
```

```python
def grado_por_nodo(G):
    grados = dict(G.degree())
    return grados

grados = grado_por_nodo(G)
print("\n6) Grado de algunos nodos:")
for n, g in list(grados.items())[:10]:
    print(f"   Nodo {n}: {g}")
```

```python
def grafo_filtrado(matriz, umbral, title):

    matriz_filtrada = np.where(matriz >= umbral, matriz, 0)
    
    G = nx.from_numpy_array(matriz_filtrada)
    
    pos = nx.spring_layout(G, seed=42)
    plt.figure(figsize=(6,6))
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray')
    plt.title(title)
    plt.show()
    
    return G
```

```python
matriz = np.array([
    [0, 0.8, 0.2, 0.0],
    [0.8, 0, 0.5, 0.1],
    [0.2, 0.5, 0, 0.9],
    [0.0, 0.1, 0.9, 0]
])

# Prueba con diferentes umbrales
G1 = grafo_filtrado(matriz, 0.0, "Grafo completo (umbral = 0.0)")
G2 = grafo_filtrado(matriz, 0.5, "Grafo filtrado (umbral = 0.5)")
G3 = grafo_filtrado(matriz, 0.8, "Grafo más estricto (umbral = 0.8)")
```

```python
G25 = grafo_filtrado(matriz, p25, 'Grafo filtrado p25')
G50 = grafo_filtrado(matriz, p50, 'Grafo filtrado p50')
G75 = grafo_filtrado(matriz, p75, 'Grafo filtrado p75')
```

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
###3
G = nx.from_numpy_array(matriz)
grados = dict(G.degree())
tamanos_nodos = [grados[n]*20 for n in G.nodes()]
```

```python
np.random.seed(42)
pos = {i: (np.random.rand(), np.random.rand(), np.random.rand()) for i in G.nodes()}
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

for i, j in G.edges():
    x = [pos[i][0], pos[j][0]]
    y = [pos[i][1], pos[j][1]]
    z = [pos[i][2], pos[j][2]]
    weight = matriz[i, j]
    ax.plot(x, y, z, c=cm.hot(weight), alpha=0.7)
```

```python
xs = [pos[i][0] for i in G.nodes()]
ys = [pos[i][1] for i in G.nodes()]
zs = [pos[i][2] for i in G.nodes()]
ax.scatter(xs, ys, zs, s=tamanos_nodos, c='lightblue', edgecolors='k')

ax.set_title("Grafo 3D con nodos proporcionales al grado y aristas coloreadas")
plt.show()
```
