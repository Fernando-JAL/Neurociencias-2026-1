```python
#Ejercicio Frecuencia de números en matriz
#Genera una matriz 6x6 de enteros aleatorios entre 1 y 10 usando NumPy.
#Recorre la matriz y guarda en un diccionario la frecuencia de cada número.
#Convierte el diccionario en una lista de tuplas (número, frecuencia) y ordénala de mayor a menor frecuencia.
#Imprime la matriz, el diccionario y la lista ordenada.

import numpy as np
matriz= np.random.randint(1,10+1,(6,6))
print(matriz)
#print(type(matriz))

frecuencia_por_num= {}
for fila in matriz:        
    for numero in fila:    
        if numero in frecuencia_por_num:
            frecuencia_por_num[numero] += 1
        else:
            frecuencia_por_num[numero] = 1
print(frecuencia_por_num)

frecuencias_tuplas = list(frecuencia_por_num.items())
#frecuencias_tuplas
frecuencias_ordenadas= sorted(frecuencias_tuplas, key=lambda x: x[1], reverse=True)
print(frecuencias_ordenadas)
```

```python
#Ejercicio Promedios por columna y gráfico de líneas
#Crea una matriz 10x5 con enteros entre 10 y 100.
#Calcula el promedio de cada columna usando un bucle.
#Guarda los resultados en un diccionario con claves "col_1" ... "col_5".
#Convierte los valores en una lista y grafícalos en una línea (line plot) mostrando el promedio por columna.

matriz= np.random.randint(10,100+1,(10,5))
filas = 10
columnas = 5
promedios = [0, 0, 0, 0, 0] 
for j in range(columnas):
    suma = 0
    for i in range(filas):
        suma += matriz[i][j]
    promedio = suma / filas
    promedios[j] = promedio 
promedios_dic = {}
for i in range(columnas):
    clave = "col_" + str(i + 1)
    promedios_dic[clave] = promedios[i]
```

```python
#Ejercicio Evolución temporal simulada
#Simula con NumPy una lista de 30 días con valores aleatorios de temperatura entre 15°C y 35°C.
#Almacena los valores en un diccionario con claves "día_1", "día_2", ....
#Usa un bucle para crear una lista con los días donde la temperatura superó 30°C.
#Grafica las temperaturas en un gráfico de línea con puntos marcados (scatter + line plot).
#Marca en rojo los días con temperatura > 30°C. 
temperaturas= np.random.randint(15,35+1,30)
temperaturas_dic = {}  # diccionario vacío
for i in range(30):
    clave = "día_" + str(i + 1)
    temperaturas_dic[clave] = temperaturas[i]
hot_days = []
for dia, temp in temperaturas_dic.items():
    if temp > 30:
        hot_days.append(dia)
        
import matplotlib.pyplot as plt
values = list(temperaturas_dic.values())
days = list(temperaturas_dic.keys())
plt.plot(values, marker='o', linestyle='-', color='r') 
plt.xticks(range(0, 30, 2), days[::2], rotation=45)  
```

```python
#Ejercicio Comparación de distribuciones
#Genera dos arreglos de 100 números aleatorios con distribución normal:
#grupo_A: media 50, desviación 10
#grupo_B: media 60, desviación 15
#Guarda en un diccionario con claves "A" y "B".
#Calcula media y desviación estándar de cada grupo y guárdalas en listas.
#Grafica ambos grupos en un histograma comparativo (superposición con colores diferentes y transparencia).

grupo_A = np.random.normal(50, 10, 100)
grupo_B = np.random.normal(60, 15, 100)
grupos = {"A": grupo_A, "B": grupo_B}
medias = []
desviaciones = []
suma = 0
for valor in grupo_A:
    suma += valor
media_A = suma / len(grupo_A)
medias.append(media_A)
# Desviación_estándar_A
suma_cuadrados = 0
for valor in grupo_A:
    suma_cuadrados += (valor - media_A) ** 2
desviacion_A = (suma_cuadrados / len(grupo_A)) ** 0.5
desviaciones.append(desviacion_A)

suma = 0
for valor in grupo_B:
    suma += valor
media_B = suma / len(grupo_B)
medias.append(media_B)
suma_cuadrados = 0
for valor in grupo_B:
    suma_cuadrados += (valor - media_B) ** 2
desviacion_B = (suma_cuadrados / len(grupo_B)) ** 0.5
desviaciones.append(desviacion_B)


plt.hist(grupo_A, color='blue', alpha=0.5, label='Grupo A')
plt.hist(grupo_B, color='red', alpha=0.5, label='Grupo B')
```

```python
#Ejercicio Clasificación de temperaturas
#Genera una Serie con 50 valores aleatorios entre -10 y 40 (temperaturas).
#Recorre la Serie y clasifica cada temperatura en un diccionario:
#"frío" si < 10
#"templado" si 10–25
#"calor" si > 25
#Convierte el diccionario en un DataFrame con columnas "Temperatura" y "Categoría".
#Cuenta cuántos valores cayeron en cada categoría.

import pandas as pd
temperaturas = np.random.randint(-10, 41, 50) 
clasificacion = {}
for i in range(len(temperaturas)):
    temp = temperaturas[i]
    if temp < 10:
        categoria = "frío"
    elif temp <= 25:
        categoria = "templado"
    else:
        categoria = "calor"
clasificacion[i] = {"Temperatura": temp, "Categoría": categoria}
dataframe = pd.DataFrame.from_dict(clasificacion, orient='index')
conteo = dataframe["Categoría"].value_counts()
```
