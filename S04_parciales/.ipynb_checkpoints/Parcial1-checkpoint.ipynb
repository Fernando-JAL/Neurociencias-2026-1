```python
#üîπ Ejercicio Frecuencia de n√∫meros en matriz

#Genera una matriz 6x6 de enteros aleatorios entre 1 y 10 usando NumPy.
import numpy as np 
import pandas as pd 
 
matriz = np.random.randint(1, 11, size=(6, 6)) 
display(matriz)

#Recorre la matriz y guarda en un diccionario la frecuencia de cada n√∫mero.
contador_1=0
contador_2=0
contador_3=0
contador_4=0
contador_5=0
contador_6=0
contador_7=0
contador_8=0
contador_9=0
contador_10=0

for i in range(6):
    for j in range(6):
        elemento = matriz[i, j]
        if elemento == 1:
            contador_1 += 1
        elif elemento == 2:
            contador_2 += 1
        elif elemento == 3:
            contador_3 += 1
        elif elemento == 4:
            contador_4 += 1
        elif elemento == 5:
            contador_5 += 1
        elif elemento == 6:
            contador_6 += 1
        elif elemento == 7:
            contador_7 += 1
        elif elemento == 8:
            contador_8 += 1
        elif elemento == 9:
            contador_9 += 1
        else:
            contador_10 += 1
            
diccionario = {
    "numero 1": contador_1,
    "numero 2": contador_2,
    "numero 3": contador_3,
    "numero 4": contador_4,
    "numero 5": contador_5,
    "numero 6": contador_6,
    "numero 7": contador_7,
    "numero 8": contador_8,
    "numero 9": contador_9,
    "numero 10": contador_10
}


#Convierte el diccionario en una lista de tuplas (n√∫mero, frecuencia) y ord√©nala de mayor a menor frecuencia.
lista_tuplas = list(diccionario.items())
display(lista_tuplas)
lista_ordenada = sorted (lista_tuplas, key=lambda x:x[1], reverse= True)
display(lista_ordenada)

df_diccionario = pd.DataFrame(list(diccionario.items()), columns=["N√∫mero", "Frecuencia"])
df_lista_ordenada = pd.DataFrame(lista_ordenada, columns=["N√∫mero", "Frecuencia"])

#Imprime la matriz, el diccionario y la lista ordenada.
display("Matriz:", matriz)
display("Diccionario de frecuencias:", df_diccionario)
display("Lista ordenada:",df_lista_ordenada)

```

```python
#üîπ Ejercicio Promedios por columna y gr√°fico de l√≠neas

#Crea una matriz 10x5 con enteros entre 10 y 100.
import numpy as np 
import pandas as pd  

matriz = np.random.randint(10,100, size=(10,5))
display(matriz)

#Calcula el promedio de cada columna usando un bucle.
diccionario= {}
for i in range(matriz.shape[1]):  
    promedio = matriz[:, i].mean() 
    diccionario[f"Columna {i+1}"] = promedio
    
display(diccionario)

#Guarda los resultados en un diccionario con claves "col_1" ... "col_5".
diccionario = {
    'col_1': diccionario['Columna 1'],
    'col_2': diccionario['Columna 2'],
    'col_3': diccionario['Columna 3'],
    'col_4': diccionario['Columna 4'],
    'col_5': diccionario['Columna 5'],
}

#Convierte los valores en una lista y graf√≠calos en una l√≠nea (line plot) mostrando el promedio por columna.
import matplotlib.pyplot as plt

list = pd.DataFrame(diccionario, index=[0]).transpose()
list.columns = ["Promedio"]
display(list)

plt.plot(list.index, list["Promedio"], marker="o", linestyle="-")
plt.title("Promedio por columna")
plt.xlabel("Columnas")
plt.ylabel("Promedio")
```

```python
#üîπ Ejercicio Evoluci√≥n temporal simulada

#Simula con NumPy una lista de 30 d√≠as con valores aleatorios de temperatura entre 15¬∞C y 35¬∞C.

import numpy as np 
import pandas as pd  
import matplotlib.pyplot as plt

Dias = np.random.randint(15,35, size=(30))
display(Dias)

#Almacena los valores en un diccionario con claves "d√≠a_1", "d√≠a_2", ....
diccionario = {}

for i in range(30):
    diccionario[f"dia {i+1}"] = Dias[i] 

display (diccionario)

#Usa un bucle para crear una lista con los d√≠as donde la temperatura super√≥ 30¬∞C.
dias_calientes = []

for i in range(len(Dias)):
    if Dias[i] > 30:
        dias_calientes.append(f"D√≠a {i+1}")

display(dias_calientes)

#Grafica las temperaturas en un gr√°fico de l√≠nea con puntos marcados (scatter + line plot).
plt.figure(figsize=(12, 5))
plt.plot(range(1, 31), Dias, marker='o', linestyle='-', label="Temperatura") 

#Marca en rojo los d√≠as con temperatura > 30¬∞C.
for i in range(len(Dias)):
    if Dias[i] > 30:
        plt.scatter(i+1, Dias[i], color='red', s=100, label=">30¬∞C" if i==0 else "")  # rojo

plt.title("Temperatura en 30 d√≠as")
plt.xlabel("D√≠a")
plt.ylabel("Temperatura (¬∞C)")
```

```python
#üîπ Ejercicio Comparaci√≥n de distribuciones

#Genera dos arreglos de 100 n√∫meros aleatorios con distribuci√≥n normal:
import  numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

#grupo_A: media 50, desviaci√≥n 10

grupo_A= np.random.normal(loc=50,scale=10, size=(100))
display (grupo_A)

#grupo_B: media 60, desviaci√≥n 15

grupo_B= np.random.normal(loc=60,scale=15, size=(100))

#Guarda en un diccionario con claves "A" y "B".
diccionario = {
    "A": grupo_A,
    "B": grupo_B
}

df = pd.DataFrame(diccionario)
display(df)

#Calcula media y desviaci√≥n est√°ndar de cada grupo y gu√°rdalas en listas.
medias = []
desviaciones = []

for area in ["A", "B"]:
    medias.append(df[area].mean())
    desviaciones.append(df[area].std())

display('medias', medias)
display('desviaciones', desviaciones)

#Grafica ambos grupos en un histograma comparativo (superposici√≥n con colores diferentes y transparencia).
plt.figure(figsize=(10,5))
plt.hist(df["A"], bins=15, alpha=0.6, label="Grupo A", color='blue')
plt.hist(df["B"], bins=15, alpha=0.6, label="Grupo B", color='green')
plt.axvline(medias[0], color='blue', linestyle='--', label="Media Grupo A")
plt.axvline(medias[1], color='green', linestyle='--', label="Media Grupo B")
plt.title("Comparaci√≥n de distribuciones")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
```

```python
#üîπ Ejercicio Clasificaci√≥n de temperaturas

#Genera una Serie con 50 valores aleatorios entre -10 y 40 (temperaturas).
import  numpy as np
import pandas as pd 

temperaturas = np.random.randint(-10,40, size=(50)) 
display(pd.DataFrame(temperaturas, columns=["Temperatura"]))


#Recorre la Serie y clasifica cada temperatura en un diccionario:
#"fr√≠o" si < 10
#"templado" si 10‚Äì25
#"calor" si > 25
diccionario = {}
for i, temp in enumerate(temperaturas, start=1):
    if temp < 10:
        categoria = "Fr√≠o"
    elif temp <= 25:
        categoria = "Templado (agustooo)"
    else:
        categoria = "Caliente"
    diccionario[f"Dia {i}"] = {"Temperatura": temp, "Categor√≠a": categoria}


#Convierte el diccionario en un DataFrame con columnas "Temperatura" y "Categor√≠a".
df = pd.DataFrame.from_dict(diccionario, orient="index")
display(df)

#Cuenta cu√°ntos valores cayeron en cada categor√≠a.
conteo = df["Categor√≠a"].value_counts()
display(conteo)
```

```python
#üîπ Ejercicio Matriz de distancias

#Genera una matriz 10x2 con coordenadas aleatorias (x, y) entre 0 y 100.
import numpy as np
import pandas as pd

coordenada= np.random.randint(0, 101, size=(10,2))
display(coordenada)

#Con un bucle, calcula las distancias euclidianas entre cada par de puntos y gu√°rdalas en un diccionario con claves (i,j).
distancias = {}  

for i in range(len(coordenada)):
    for j in range(len(coordenada)):
        xi, yi = coordenada[i]
        xj, yj = coordenada[j]
        distancia = np.sqrt((xi - xj)**2 + (yi - yj)**2)
        distancias[(i,j)] = distancia
display(distancia)

#Convierte el diccionario en un DataFrame de 10x10 (matriz de distancias).

df_distancias = pd.DataFrame(np.zeros((10,10)))

for key, value in distancias.items():
    i, j = key
    df_distancias.iloc[i,j] = value

display(df_distancias)


#Encuentra el par de puntos m√°s cercanos y el m√°s lejano
df_sin_diagonal = df_distancias.copy()
np.fill_diagonal(df_sin_diagonal.values, np.nan)
min_valor = df_sin_diagonal.min().min()
min_pos = np.where(df_sin_diagonal == min_valor)
par_cercano = (min_pos[0][0], min_pos[1][0])
max_valor = df_sin_diagonal.max().max()
max_pos = np.where(df_sin_diagonal == max_valor)
par_lejano = (max_pos[0][0], max_pos[1][0])

print(f"Par m√°s cercano: {par_cercano} con distancia {min_valor:.2f}")
print(f"Par m√°s lejano: {par_lejano} con distancia {max_valor:.2f}")
```

```python
#üîπ Ejercicio An√°lisis de series temporales}

#Genera con NumPy una Serie de 365 valores aleatorios entre 10 y 40 (temperatura diaria de un a√±o).
import numpy as np
import pandas as pd

temperaturas = np.random.randint(10, 41, size=(365))
display(temperaturas)

#Convierte en un DataFrame con columnas "D√≠a" y "Temperatura".
df = pd.DataFrame({
    "D√≠a": range(1, 366),
    "Temperatura": temperaturas
})
display(df)
#Usa un bucle para calcular la media de cada mes (asume 30 d√≠as por mes).
#Guarda las medias en un diccionario {Mes: Promedio}.
medias_mensuales = {}  

for mes in range(12): 
    inicio = mes * 30
    fin = inicio + 30
    promedio = df["Temperatura"].iloc[inicio:fin].mean()
    medias_mensuales[f"Mes {mes+1}"] = promedio

df_medias = pd.DataFrame(list(medias_mensuales.items()), columns=["Mes", "Promedio"])
display(df_medias)

#Convierte el diccionario en una Series y ord√©nala de mayor a menor.
series_mensuales = pd.Series(medias_mensuales)
series_mensuales = series_mensuales.sort_values(ascending=False)
display(series_mensuales)
```

```python
#üîπ Ejercicio Actividad neuronal por regiones
# Genera una matriz 100x5 con NumPy representando la actividad de 100 neuronas en 5 regiones cerebrales (valores entre 0 y 1 como tasas de disparo).

import numpy as np
import pandas as pd

#np.random.seed(123)

matriz = np.random.uniform(0, 1, size=(100, 5))

# Convierte la matriz en un DataFrame con columnas "Regi√≥n_1"... "Regi√≥n_5".
diccionario = {
    "Region_1" : matriz[:,0],
    "Region_2" : matriz[:,1],
    "Region_3" : matriz[:,2],
    "Region_4" : matriz[:,3],
    "Region_5" : matriz[:,4]
}

df = pd.DataFrame(diccionario)
display(df)

# Usa un bucle para calcular la actividad media de cada regi√≥n y gu√°rdala en un diccionario.

regiones = ["Region_1", "Region_2", "Region_3", "Region_4", "Region_5"]


for region in regiones:
    mean = df[region].mean()
    diccionario[region] = mean
    
display(diccionario)
    
# Convierte el diccionario en una Series y ord√©nala de mayor a menor.

series = pd.DataFrame(diccionario, index=[0]).transpose()
series.columns = ["Promedio"]
series = series.sort_values(by="Promedio", ascending=False)
series
```

```python
#üîπ Ejercicio Conectividad funcional (correlaciones)
#Simula con NumPy una matriz 200x4 con series de tiempo de actividad fMRI en 4 √°reas cerebrales.
import numpy as np
import pandas as pd
matriz = np.random.uniform(size=(200, 4))

#Convierte en un DataFrame con columnas "√Årea_1"... "√Årea_4".
diccionario = {
    "Area_1" : matriz[:,0],
    "Area_2" : matriz[:,1],
    "Area_3" : matriz[:,2],
    "Area_4" : matriz[:,3]
}

df = pd.DataFrame(diccionario)
display(df)

#Con un bucle, calcula la correlaci√≥n de Pearson entre todas las parejas de√°reas y gu√°rdala en un diccionario con claves (√Årea_i, √Årea_j).

areas = ["Area_1","Area_2","Area_3","Area_4"]
diccionario_corr={}

for area_i in areas:
    for area_j in areas:
        corelacion= df[area_i].corr(df[area_j])
        diccionario_corr[area_i, area_j] = corelacion
    
display(diccionario_corr)

#Convierte el diccionario en un DataFrame cuadrado (matriz de conectividad).
df_corr = pd.Series(diccionario_corr).unstack()
display(df_corr)


```

```python
#üîπ Ejercicio Potenciales de acci√≥n por neurona
#Genera una lista de 50 neuronas.
Neuronas = list(range(50))

#Para cada neurona, simula con NumPy una Serie de 100 ms de actividad (valores enteros 0 o 1 indicando si dispar√≥ o no).

import numpy as np
import pandas as pd

Disparos = np.random.randint(0, 2, size=(100,50))
display(Disparos)

#Guarda los resultados en un diccionario {Neurona: Serie}.  

diccionario = {}

for i in range(50):
    diccionario[f"Neurona {i+1}"] = Disparos[:,i]


    
#Crea un DataFrame con todas las neuronas (filas) y tiempos (columnas).
df = pd.DataFrame(diccionario)
display(df)

#Usa un bucle para contar cu√°ntos disparos totales tuvo cada neurona y genera una Series con esos valores.

for i in Neuronas:
    sum = df[f"Neurona {i+1}"].sum()
    diccionario[f"Neurona {i+1}"] = sum
    
series = pd.DataFrame(diccionario, index=range(1)).transpose()
series.columns = ["Total de disparos"]

display(series)
```

```python
#üîπ Ejercicio EEG por canales

#Simula una matriz 500x8 con NumPy representando 500 muestras de se√±al EEG en 8 canales (valores flotantes entre -100 y 100 ¬µV).
import numpy as np
import pandas as pd

matriz = np.random.uniform(-100, 100, size=(500, 8))


#Convierte la matriz en un DataFrame con columnas "Canal_1"... "Canal_8".
diccionario = {
    "canal_1" : matriz[:,0],
    "canal_2" : matriz[:,1],
    "canal_3" : matriz[:,2],
    "canal_4" : matriz[:,3],
    "canal_5" : matriz[:,4],
    "canal_6" : matriz[:,5],
    "canal_7" : matriz[:,6],
    "canal_8" : matriz[:,7]
}

df = pd.DataFrame(diccionario)
display(df)

#Usa un bucle para calcular la potencia media de cada canal (media del cuadrado de la se√±al).
canales = ["canal_1","canal_2","canal_3","canal_4","canal_5","canal_6","canal_7","canal_8"]

potencias = {}

for canal in canales:
    potencia = (df[canal] ** 2).mean()
    potencias[canal] = potencia
    
print(potencias)
    
#Guarda los resultados en un diccionario y convi√©rtelo en una Series ordenada.
series = pd.DataFrame(potencias, index=[0]).transpose()
series.columns = ["Potencia promedio"]
series = series.sort_values(by="Potencia promedio", ascending=False)
display(series)

#Identifica el canal con mayor potencia.
print("Canal con mayor potencia promedio")
display(series.iloc[0,:])



```

```python
#üîπ Ejercicio An√°lisis de respuestas a est√≠mulos


#Genera un DataFrame con 30 sujetos (filas) y 3 condiciones experimentales ("Cond_A", "Cond_B", "Cond_C") con valores de tiempo de reacci√≥n simulados (valores entre 200‚Äì800 ms).
import numpy as np
import pandas as pd

Sujetos = np.random.uniform(200, 800, size=(30, 3))

display(Sujetos)

diccionario = {
    "Cond_A" : Sujetos[:,0],
    "Cond_B" : Sujetos[:,1],
    "Cond_C" : Sujetos[:,2],
}

df = pd.DataFrame(diccionario)

#Usa un bucle para calcular el promedio por condici√≥n en toda la muestra.

condiciones = ["Cond_A","Cond_B","Cond_C" ]

for condicion in condiciones:
    mean = df[condicion].mean()
    diccionario[condicion] = mean
    
display(diccionario)

#Convierte los resultados en un diccionario y luego en una Series.
series = pd.DataFrame(diccionario, index=[0]).transpose()
series.columns = ["Promedio"]   # Solo una columna con los promedios
series = series.sort_values(by="Promedio", ascending=False)
display(series)

#Agrega una columna "Mejor_condici√≥n" al DataFrame indicando la condici√≥n m√°s r√°pida por sujeto.
df["Mejor_condici√≥n"] = df[condiciones].idxmin(axis=1)

#Cuenta cu√°ntos sujetos respondieron mejor en cada condici√≥n.
conteo = df["Mejor_condici√≥n"].value_counts()
conteo
```
