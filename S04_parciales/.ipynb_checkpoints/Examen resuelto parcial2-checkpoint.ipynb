```python
import numpy as np
import networkx as nx
from networkx.algorithms import community
import scipy.io
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Ejercicio 1
Coactivation_matrix = r"/Users/sofish/Documents/GitHub/Neurociencias-2026-1/S03_datasets/BCT/Coactivation_matrix.mat"
data = scipy.io.loadmat(Coactivation_matrix)
data

```

```python
(data.keys())
```

```python
matriz = data["Coactivation_matrix"]
matriz = np.triu(matriz)
matriz = (matriz + matriz.T) 
np.fill_diagonal(matriz, 0)
matriz
#Red aleatoria
```

```python
#Crear grafo
G = nx.from_numpy_array(matriz)
G
print("Número de nodos:", G.number_of_nodes())
("Número de aristas:", G.number_of_edges())
```

```python
cluster_coeff = nx.average_clustering(G)

if nx.is_connected(G):
    L = nx.average_shortest_path_length(G)
else:
    largest_cc = max(nx.connected_components(G), key=len)
    G0 = G.subgraph(largest_cc)
    L = nx.average_shortest_path_length(G0)
print("Clustering",cluster_coeff,"Longitud",L)
```

```python
#Coeficiente de mundo pequeño
try:
    path_length = nx.average_shortest_path_length(G)
except nx.NetworkXError:
    path_length = np.nan

G_rand = nx.gnm_random_graph(n=G.number_of_nodes(), m=G.number_of_edges())
clust_rand = nx.average_clustering(G_rand)
path_rand = nx.average_shortest_path_length(G_rand)
small_world_sigma = (cluster_coeff / clust_rand) / (path_length / path_rand)
(path_length,cluster_coeff, clust_rand, path_rand) 
print("Coeficiente de mundo pequeño",small_world_sigma)
```

```python
#Comunidades
from networkx.algorithms import community
comunidades = community.greedy_modularity_communities(G)
modularity = community.modularity(G, comunidades)
(comunidades, modularity)
```

```python
#Hub
degree_dict = dict(G.degree())
betweenness = nx.betweenness_centrality(G)
degree_dict
```

```python
betweenness
```

```python
#Eficiencia global y local
```

```python
global_eff = nx.global_efficiency(G)
local_eff = nx.local_efficiency(G)
print(global_eff)
print(local_eff)
```

```python
#Grado de cada nodo
grados = dict(G.degree())
for nodo, grado in grados.items():
    print(f"Nodo {nodo}: {grado}")
```

```python
#Correlación de Pearson- En este caso se utilizó esta matriz
matriz = data['Coactivation_matrix']
df = pd.DataFrame(matriz)
df.to_csv("connectivity_coactivation.csv", index=False)

corr_matrix = np.corrcoef(matriz)
corr_df = pd.DataFrame(corr_matrix)
corr_df.to_csv("connectivity_correlation.csv", index=False)
(corr_matrix)
print(corr_df.to_csv)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, cmap="hot")

```

```python
#Percentiles
p25 = np.percentile(matriz, 25)
p50 = np.percentile(matriz, 50)
p75 = np.percentile(matriz, 75)
print(f"Percentil 25%: {p25:.4f}, 50%: {p50:.4f}, 75%: {p75:.4f}")

percentiles = [p25, p50, p75]
for i, perc in enumerate(percentiles, start=1):
    matriz_filtro = np.where(matriz >= perc, matriz, 0)
    G = nx.from_numpy_array(matriz_filtro)
    pos = nx.spring_layout(G, seed=42)
    
    plt.figure(figsize=(10, 10))
    nx.draw_networkx(G, pos, with_labels=True, node_color="yellow", edge_color="black", node_size=500)
    plt.title(f"Grafo filtrado - Percentil {int([25,50,75][i-1])} ({perc:.4f})")
    plt.show()

```

```python
#Tamaño del nodo
degree_dict = dict(G.degree(weight="weight"))
node_sizes = [v * 50 for v in degree_dict.values()]  
np.random.seed(42)
n_nodos = matriz.shape[0]
coordenadas_df = pd.DataFrame({
    "x": np.random.rand(n_nodos),
    "y": np.random.rand(n_nodos),
    "z": np.random.rand(n_nodos)
})
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection="3d")
for node in G.nodes():
    x, y, z = coordenadas_df.loc[node, ["x","y","z"]]
    ax.scatter(x, y, z, s=node_sizes[node], c="yellow", alpha=0.8)
    ax.text(x, y, z+0.02, str(node), fontsize=8, ha="center")

# Aristas según peso (map Hot)
weights_dict = nx.get_edge_attributes(G, "weight")
min_w = min(weights_dict.values())
max_w = max(weights_dict.values())
cmap = plt.get_cmap('hot')

for i, j, data in G.edges(data=True):
    x_vals = [coordenadas_df.loc[i, "x"], coordenadas_df.loc[j, "x"]]
    y_vals = [coordenadas_df.loc[i, "y"], coordenadas_df.loc[j, "y"]]
    z_vals = [coordenadas_df.loc[i, "z"], coordenadas_df.loc[j, "z"]]
    normalized_weight = (data["weight"] - min_w) / (max_w - min_w)
    color = cmap(normalized_weight)
    ax.plot(x_vals, y_vals, z_vals, c=color, linewidth=2, alpha=0.7)

plt.show()
```

```python
#Ejercicio 4
from networkx.algorithms import community
from scipy.signal import coherence, hilbert
import mne
chb01_path = r"/Users/sofish/Documents/GitHub/Neurociencias-2026-1/S03_datasets/EEG_practica/chb01_01.edf"
sfreq_target = 256       
fmin, fmax = 13, 30 

raw = mne.io.read_raw_edf(chb01_path, preload=True)
raw.resample(sfreq_target)
raw.filter(1., 45., fir_design="firwin")
chb01_data = raw.get_data()           
ch_names = raw.info["ch_names"]
n_channels = len(ch_names)
```

```python
#Mundo pequeño
coh_matrix = np.zeros((n_channels, n_channels))

for i in range(n_channels):
    for j in range(n_channels):
        f, Cxy = coherence(chb01_data[i], chb01_data[j], fs=sfreq_target, nperseg=sfreq_target*2)
        mask = (f >= fmin) & (f <= fmax)
        coh_matrix[i, j] = np.mean(Cxy[mask])
```

```python
G = nx.from_numpy_array(coh_matrix)
clust_coeff = nx.average_clustering(G, weight="weight")
```

```python
try:
    path_length = nx.average_shortest_path_length(G, weight="weight")
except nx.NetworkXError:
    path_length = np.nan  
G_rand = nx.gnm_random_graph(n=G.number_of_nodes(), m=G.number_of_edges())
clust_rand = nx.average_clustering(G_rand)
path_rand = nx.average_shortest_path_length(G_rand)
small_world_sigma = (clust_coeff / clust_rand) / (path_length / path_rand)
(path_length, clust_coeff, clust_rand, path_rand)
print("Coeficiente de mundo pequeño:", small_world_sigma)
```

```python
#Comunidades
communities = community.greedy_modularity_communities(G, weight="weight")
modularity = community.modularity(G, communities, weight="weight")
(communities, modularity)
```

```python
#Hub
degree_dict = dict(G.degree(weight="weight"))
betweenness = nx.betweenness_centrality(G, weight="weight")
(degree_dict, betweenness)
```

```python
#Eficiencia global y local
global_eff = nx.global_efficiency(G)
local_eff = nx.local_efficiency(G)
print("Eficiencia global:", global_eff)
print("Eficiencia local:", local_eff)
```

```python
#Grado por cada nodo
grado_por_nodo = dict(G.degree(weight="weight"))
for nodo, grado in grado_por_nodo.items():
    print(f"Nodo {nodo}: {grado}")

```
