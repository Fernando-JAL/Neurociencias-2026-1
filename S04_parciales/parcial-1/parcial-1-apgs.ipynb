# 1. Frecuencia de números en matriz

a) Genera una matriz 6x6 de enteros random entre 1 y 10 usando NumPy.

```python
matriz = np.random.randint(1, 11, size = (6,6))
print("Matriz:", matriz)
```

b) Recorre la matriz y guarda en un diccionario la frecuencia de cada número.

```python
frecuencia = {} # Diccionario para guardar las frecuencias
for fila in matriz:
    for numero in fila:
        if numero in frecuencia:
            frecuencia[numero] += 1 # Si ya existe, sumar 1
        else: 
            frecuencia[numero] = 1 # Si no existe, inicializar en 1
print("Frecuencia de cada número:", frecuencia)
```

c) Convierte el diccionario en una lista de tuplas (número, frecuencia) y ordénala de mayor a menor frecuencia.

```python
lista_ordenada = sorted(frecuencia.items(), key = lambda x: x[1], reverse = True) 
# frecuencia.items() convierte diccionario → lista tipo (número, frecuencia)
# key = lambda x: x[1] indica a sorted() que ordene el 2° elemento de c/tupla
print("Lista de frecuencias ordenada:", lista_ordenada)
```

# 2. Promedios por columna y gráfico de líneas

a) Crea una matriz 10x5 con enteros entre 10 y 100.

```python
matriz = np.random.randint(10, 101, size = (10, 5))
matriz
```

b) Calcula el promedio de cada columna usando un bucle.

c) Guarda los resultados en un diccionario con claves "col_1" ... "col_5".

```python
promedios = {} # Diccionario para guardar {col_: promedio}
for i in range(5): # Recorrer las 5 columnas
    suma = 0 
    for j in range(10): # Recorrer las 10 filas
        suma += matriz[j][i] # Sumar elementos columna actual
    promedio = suma / 10
    nombre_columna = "col_" + str(i+1) # Nombrar cada columna 
    promedios[nombre_columna] = promedio 
print("Promedios por columna:", promedios)
```

d) Convierte los valores en una lista y grafícalos en una línea (line plot) mostrando el promedio por columna.

```python
lista = list(promedios.values()) # Extraer promedios pro columna y listarlos
plt.plot(lista, marker = "o")
plt.title("Promedio por columna")
plt.xlabel("Columna")
plt.ylabel("Promedio")
plt.xticks(range(5), list(promedios.keys()))
plt.show()
```

# 3. Evolución temporal simulada

a) Simula con NumPy una lista de 30 días con valores aleatorios de temperatura entre 15°C y 35°C.

```python
temperaturas = np.random.randint(15, 35, size = (30))
print("Temperaturas aleatorias:", temperaturas)
```

b) Almacena los valores en un diccionario con claves "día_1", "día_2", ....b) 

```python
dias = {} # Diccionario para almacenar temperaturas por día
for i in range(30):
    nombre_dia = "dia_" + str(i+1)
    dias[nombre_dia] = temperaturas[i]
print("Asignación de temperaturas aleatorias a días:", dias)
```

c) Usa un bucle para crear una lista con los días donde la temperatura superó 30°C.

```python
dias_calurosos = [] # Lista para almacenar días donde la temperatura > 30°C
for nombre_dia in dias:
    if dias[nombre_dia] > 30:
        dias_calurosos.append(nombre_dia) # Guardar el índice del día
print("Días donde la temperatura > 30°C:", dias_calurosos)
```

d) Grafica las temperaturas en un gráfico de línea con puntos marcados (scatter + line plot).

e) Marca en rojo los días con temperatura > 30°C.

```python
plt.plot(temperaturas, linestyle = "-")
plt.scatter(range(30), temperaturas)

for nombre_dia in dias_calurosos:
    indice = list(dias.keys()).index(nombre_dia) 
    plt.scatter(indice, dias[nombre_dia], color="red")

plt.title("Temperatura diaria (°C)")                       
plt.xlabel("Día")
plt.ylabel("Temperatura")
plt.show()
```

# 5. Clasificación de temperaturas

a) Genera una serie con 50 valores aleatorios entre -10 y 40 (temperaturas).

```python
temperaturas = pd.Series(np.random.randint(-10, 41, size=50))
print("Serie de temperaturas aleatorias:", temperaturas)
```

b) Recorre la serie y clasifica cada temperatura en un diccionario:
- "frío" si < 10
- "templado" si 10–25
- "calor" si > 25


```python
clasificacion = {}  # Diccionario para almacenar {temperatura: categoría}
for i in range(len(temperaturas)):
    temp = temperaturas[i]
    if temp < 10:
        categoria = "frío"
    elif temp <= 25:
        categoria = "templado"
    else:
        categoria = "calor"
    clasificacion[i] = {"Temperatura": temp, "Categoría": categoria}  
clasificacion
```

c) Convierte el diccionario en un DataFrame con columnas "Temperatura" y "Categoría".

```python
df = pd.DataFrame.from_dict(clasificacion, orient="index")
print("\nTemperatura y categoría por fila:")
print(df)
```

d) Cuenta cuántos valores cayeron en cada categoría.

```python
conteo = df["Categoría"].value_counts()
conteo
```

# 6. Matriz de distancias

a) Genera una matriz 10x2 con coordenadas aleatorias (x, y) entre 0 y 100.

```python
coordenadas = np.random.randint(0, 101, size=(10, 2))
print("Matriz de coordenadas (x, y):", coordenadas)
```

b) Con un bucle, calcula las distancias euclidianas entre cada par de puntos y guárdalas en un diccionario con claves (i,j).

```python
distancias = {}  # Diccionario para almacenar {(x,y): distancia entre puntos}
for i in range(10):
    for j in range(10):
        punto1 = coordenadas[i]
        punto2 = coordenadas[j]
        distancia = np.sqrt((punto1[0] - punto2[0])**2 + (punto1[1] - punto2[1])**2)
        distancias[(i, j)] = distancia
distancias
```

c) Convierte el diccionario en un DataFrame de 10x10 (matriz de distancias).


```python
matriz = pd.DataFrame(np.zeros((10, 10)))  # Matriz vacía
for (i, j), valor in distancias.items():
    matriz.iloc[i, j] = valor
matriz
```

d) Encuentra el par de puntos más cercanos y el más lejano.

```python
distancias_filtradas = {k: v for k, v in distancias.items() if k[0] != k[1]}

mas_cercanos = min(distancias_filtradas, key=distancias_filtradas.get)
mas_lejanos = max(distancias_filtradas, key=distancias_filtradas.get)

print("Par de puntos más cercanos:", mas_cercanos)
print("Par de puntos más lejanos:", mas_lejanos)
```

# 7. Análisis de series temporales

a) Genera con NumPy una Serie de 365 valores aleatorios entre 10 y 40 (temperatura diaria de un año).

```python
temperaturas = np.random.randint(10, 41, size = 365)
temperaturas # Temperaturas en 365 días
```

b) Convierte en un DataFrame con columnas "Día" y "Temperatura".

```python
df = pd.DataFrame({
    "Día del año": range(1, 366),
    "Temperatura": temperaturas
})
df
```

c) Usa un bucle para calcular la media de cada mes (asume 30 días por mes).

d) Guarda las medias en un diccionario {Mes: Promedio}.

```python
medias_mensuales = {}  # Diccionario para guardar {mes: promedio}
for mes in range(1, 13):  # 12 meses
    inicio = (mes - 1) * 30
    fin = inicio + 30
    promedio = df["Temperatura"].iloc[inicio:fin].mean()
    medias_mensuales[mes] = promedio
medias_mensuales
```

e) Convierte el diccionario en una Series y ordénala de mayor a menor.

```python
serie_ordenada = pd.Series(medias_mensuales).sort_values(ascending=False)
print("\nPromedios mensuales ordenados de mayor a menor:")
print(serie_ordenada)
```

# 9. Conectividad funcional (correlaciones)

a) Simula con NumPy una matriz 200x4 con series de tiempo de actividad fMRI en 4 áreas cerebrales.

```python
actividad_fmri = np.random.rand(200, 4)
actividad_fmri
```

b) Convierte en un DataFrame con columnas "Área_1"... "Área_4".

```python
df = pd.DataFrame(actividad_fmri, columns=["Área_1", "Área_2", "Área_3", "Área_4"])
print("Actividad fMRI por área:")
print(df.head())
```

c) Con un bucle, calcula la correlación de Pearson entre todas las parejas de áreas y guárdala en un diccionario con claves (Área_i, Área_j).

```python
correlaciones = {}  # Diccionario para guardar {(Área_i, Área_j): correlación}
columnas = df.columns
for i in range(len(columnas)):
    for j in range(len(columnas)):
        area_i = columnas[i]
        area_j = columnas[j]
        r = df[area_i].corr(df[area_j])  # Correlación de Pearson
        correlaciones[(area_i, area_j)] = r
correlaciones
```

d) Convierte el diccionario en un DataFrame cuadrado (matriz de conectividad).

```python
matriz_conectividad = pd.DataFrame(index=columnas, columns=columnas)
for (area_i, area_j), valor in correlaciones.items():
    matriz_conectividad.loc[area_i, area_j] = valor
print("\nMatriz de conectividad funcional (correlaciones):")
print(matriz_conectividad)
```

# 10. Potenciales de acción por neurona

a) Genera una lista de 50 neuronas.

```python
neuronas = [f"Neurona_{i+1}" for i in range(50)]
neuronas
```

b) Para cada neurona, simula con NumPy una Serie de 100 ms de actividad (valores enteros 0 o 1 indicando si disparó o no). 
c) Guarda los resultados en un diccionario {Neurona: Serie}.

```python
actividad = {}  # Diccionario para guardar {Neurona: Serie}
for nombre in neuronas:
    disparos = np.random.randint(0, 2, size=100)  # 0 o 1, tamaño 100
    actividad[nombre] = disparos
actividad
```

d) Crea un DataFrame con todas las neuronas (filas) y tiempos (columnas).

```python
df = pd.DataFrame.from_dict(actividad, orient="index")
df.columns = [f"ms_{i+1}" for i in range(100)]  # Nombrar columnas como ms_1, ms_2, ..., ms_100
df
```

e) Usa un bucle para contar cuántos disparos totales tuvo cada neurona y genera una Series con esos valores.

```python
conteo_disparos = {}  # Diccionario para guardar {Neurona: Total de disparos}
for nombre in df.index:
    total = df.loc[nombre].sum()
    conteo_disparos[nombre] = total
serie_disparos = pd.Series(conteo_disparos)
print("\nDisparos totales por neurona:")
print(serie_disparos)
```

# 11. EEG por canales

a) Simula una matriz 500x8 con NumPy representando 500 muestras de señal EEG en 8 canales (valores flotantes entre -100 y 100 µV).

```python
eeg = np.random.uniform(-100, 100, size=(500, 8))
eeg
```

b) Convierte la matriz en un DataFrame con columnas "Canal_1"... "Canal_8".

```python
df = pd.DataFrame(eeg, columns=[f"Canal_{i+1}" for i in range(8)])
df
```

c) Usa un bucle para calcular la potencia media de cada canal (media del cuadrado de la señal).

d) Guarda los resultados en un diccionario y conviértelo en una serie ordenada.

```python
potencias = {} # Diccionario para almacenar {canal_: potencia media}
for canal in df.columns:
    potencia = (df[canal] ** 2).mean()
    potencias[canal] = potencia
serie_potencias = pd.Series(potencias).sort_values(ascending=False)
print("\nPotencia media por canal (ordenada):")
print(serie_potencias)
```

e) Identifica el canal con mayor potencia.

```python
canal_maximo = serie_potencias.idxmax()
print(f"\nCanal con mayor potencia: {canal_maximo}")
```

# 12. Análisis de respuestas a estímulos

a) Genera un DataFrame con 30 sujetos (filas) y 3 condiciones experimentales ("Cond_A", "Cond_B", "Cond_C") con valores de tiempo de reacción simulados (valores entre 200–800 ms).

```python
datos = np.random.randint(200, 801, size=(30, 3)) 
df = pd.DataFrame(datos, columns=["Cond_A", "Cond_B", "Cond_C"])
df
```

b) Usa un bucle para calcular el promedio por condición en toda la muestra.
c) Convierte los resultados en un diccionario y luego en una serie.

```python
promedios = {} # Diccionario para almacenar {Cond_: promedio}
for columna in df.columns:
    promedio = df[columna].mean()
    promedios[columna] = promedio
serie_promedios = pd.Series(promedios)
print("\nPromedio por condición:")
print(serie_promedios)
```

d) Agrega una columna "Mejor_condición" al DataFrame indicando la condición más rápida por sujeto.


```python
df["Mejor_condición"] = df[["Cond_A", "Cond_B", "Cond_C"]].idxmin(axis=1)
df
```

e) Cuenta cuántos sujetos respondieron mejor en cada condición.

```python
conteo_mejor = df["Mejor_condición"].value_counts()
print("\nConteo de sujetos por mejor condición:")
print(conteo_mejor)
```
