```python
from pathlib import Path
import shutil
```

```python
#Organizar archivos por ensayos
def get_files_by_run(root_path):
    files_by_run = {f"ensayo{r:02d}": [] for r in range(1, 15)}
    for edf_file in Path(root_path).glob("S*R*.edf"):
        run_num = int(edf_file.stem.split("R")[1][:2])
        files_by_run[f"ensayo{run_num:02d}"].append(edf_file)
    return files_by_run

# Ejemplo de uso
file_path = r"C:\Users\mafig\Downloads\eeg-motor-movementimagery-dataset-1.0.0"
ensayo= get_files_by_run(file_path)
print(ensayo["ensayo03"])
```

```python
def get_files_by_run(root_path):
    root = Path(root_path)
    files_by_run = {f"run{r:02d}": [] for r in range(1, 15)}

    # Busca recursivamente en todas las carpetas de sujetos
    for edf_file in root.rglob("S*R*.edf"):
        run_part = edf_file.stem.split("R")[1][:2]  # Ej. "01"
        run_num = int(run_part)
        files_by_run[f"run{run_num:02d}"].append(edf_file)

    return files_by_run

# Ejemplo de uso:
root = r"C:\Users\mafig\Downloads\eeg-motor-movementimagery-dataset-1.0.0\files"
runs = get_files_by_run(root)

# Verifica archivos del run 3 (por ejemplo)
print("Archivos del run03:")
for f in runs["run03"]:
    print(f)
```
