```python

import numpy as np
import pandas as pd
import networkx as nx
import seaborn as sns
import matplotlib.pyplot as plt
from mi_modulo import * 
np.random.seed(123)

ensayo5_excel = r"C:\Users\shade\OneDrive\Documents\GitHub\Neurociencias-2026-1\eeg-motor-movementimagery_Ensayo_05.xlsx"
ensayo6_excel = r"C:\Users\shade\OneDrive\Documents\GitHub\Neurociencias-2026-1\eeg-motor-movementimagery_Ensayo_06.xlsx"
coordenadas_csv = r"C:\Users\shade\OneDrive\Documents\GitHub\Neurociencias-2026-1\MI_coordinates.csv"


ensayo5_dict = pd.read_excel(ensayo5_excel, sheet_name=None, index_col=0)
ensayo6_dict = pd.read_excel(ensayo6_excel, sheet_name=None, index_col=0)

print("Sujetos en Ensayo 5:", ensayo5_dict.keys())
print("Sujetos en Ensayo 6:", ensayo6_dict.keys())

ensayos5_dfs = [ensayo5_dict[s] for s in ensayo5_dict.keys()]
ensayos6_dfs = [ensayo6_dict[s] for s in ensayo6_dict.keys()]
```

```python
resultado5_df = capsula(ensayos5_dfs, 0.5, 0.6)   
resultado6_df = capsula(ensayos6_dfs, 0.4, 0.7)   

print("Matriz grupal Ensayo 5:")
display(resultado5_df)

print("Matriz grupal Ensayo 6:")
display(resultado6_df)
```

```python
ensayos_dfs = ensayos5_dfs  

th25, th50, th75 = [], [], []
promedio, varianza = [], []
mu_sigma = []

fig, axes = plt.subplots(2, 2, figsize=(12, 7))

for sujeto in ensayos_dfs:
    data = sujeto.stack().values

    th25.append(np.percentile(data, 25))
    th50.append(np.percentile(data, 50))
    th75.append(np.percentile(data, 75))
    promedio.append(np.mean(data))
    varianza.append(np.var(data))
    mu_sigma.append(np.mean(data) + np.var(data))

    sns.histplot(data, ax=axes[0][0], kde=True)
    sns.ecdfplot(data, ax=axes[0][1])

axes[1][0].plot(th25, label='Percentil 25%')
axes[1][0].plot(th50, label='Percentil 50%')
axes[1][0].plot(th75, label='Percentil 75%')
axes[1][0].plot(promedio, label='Media')
axes[1][0].plot(varianza, label='Varianza')
axes[1][0].plot(mu_sigma, label='Media + Varianza')
axes[1][0].legend()

plt.suptitle('Estadísticos descriptivos — Ensayo 5')
plt.tight_layout()
plt.show()
```

```python
from scipy import stats

estadisticos_df = pd.DataFrame(
    columns=['th25', 'th50', 'th75', 'promedio', 'mu_sigma'],
    index=['min', 'max', 'mediana', 'MAD']
)

columnas = [th25, th50, th75, promedio, mu_sigma]

min_, max_, mediana_, cosa_ = [], [], [], []

for columna in columnas:
    d1 = np.min(columna)
    d2 = np.max(columna)
    d3 = np.median(columna)
    d4 = stats.median_abs_deviation(columna)

    min_.append(d1)
    max_.append(d2)
    mediana_.append(d3)
    cosa_.append(d4)

estadisticos_df.loc['min'] = min_
estadisticos_df.loc['max'] = max_
estadisticos_df.loc['mediana'] = mediana_
estadisticos_df.loc['MAD'] = cosa_

display(estadisticos_df)
```
