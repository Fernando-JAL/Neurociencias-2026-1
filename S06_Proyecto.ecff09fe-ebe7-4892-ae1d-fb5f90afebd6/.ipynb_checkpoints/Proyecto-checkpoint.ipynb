```python
# los unix timestamps están desplazados en fecha pero no en hora.
# convertir el formato Empatica a series temporales correctas.
def to_timeseries(df: pd.DataFrame, unit: str) -> pd.DataFrame:
    t = df.iloc[:, 0]
    vals = df.iloc[:, 1:].copy()
    vals.index = pd.to_datetime(t, unit=unit)  # unit='s' o 'ms' según info.txt
    return vals
```

```python
# determinar unidades
def guess_timestamp_unit(series: pd.Series) -> str:
    # Heurística: timestamps ~1e9 -> segundos; ~1e12 -> milisegundos
    m = series.iloc[:100].median()
    return "ms" if m > 1e11 else "s"
```

```python
TARGET_FS = 4.0  # Hz (frecuencia común moderada y suficiente para combinar distintas Fs)

def butter_bandpass(low, high, fs, order=4):
    b, a = sp.butter(order, [low/(fs/2), high/(fs/2)], btype='band')
    return b, a

def butter_lowpass(cut, fs, order=4):
    b, a = sp.butter(order, cut/(fs/2), btype='low')
    return b, a

def apply_filter(x: np.ndarray, b, a) -> np.ndarray:
    return sp.filtfilt(b, a, x, axis=0)

def resample_to(df: pd.DataFrame, fs_target=TARGET_FS) -> pd.DataFrame:
    # Asegurar DatetimeIndex y periodo uniforme
    rule = pd.Timedelta(seconds=1.0/fs_target)
    return df.resample(rule).mean().interpolate(method="linear", limit=8)  # límite acotado y reproducible

def preprocess_modalities(raw: dict, fs_guess: dict) -> pd.DataFrame:
    # raw: dict nombre->DataFrame (EDA,BVP,HR,IBI,ACC,TEMP)
    # fs_guess: frecuencias originales desde info.txt (dict)
    # 1) resamplear cada señal a TARGET_FS
    ts = {}
    for name, df in raw.items():
        # guess timestamp unit
        unit = guess_timestamp_unit(df.iloc[:, 0])
        vals = to_timeseries(df, unit=unit)
        # resample primero (reduce costo)
        vals_r = resample_to(vals, TARGET_FS)
        # filtros por modalidad
        if name == "EDA":
            b, a = butter_lowpass(1.0, TARGET_FS, order=4)
            vals_r[:] = apply_filter(vals_r.values, b, a)
        elif name == "BVP":
            # en 4 Hz no tiene sentido una banda alta; suavizado leve
            b, a = butter_lowpass(1.5, TARGET_FS, order=4)
            vals_r[:] = apply_filter(vals_r.values, b, a)
        elif name == "HR":
            b, a = butter_lowpass(0.5, TARGET_FS, order=4)
            vals_r[:] = apply_filter(vals_r.values, b, a)
        elif name == "IBI":
            vals_r[:] = pd.Series(vals_r.values.ravel()).rolling(5, min_periods=1).mean().values.reshape(-1, 1)
        elif name == "TEMP":
            b, a = butter_lowpass(0.1, TARGET_FS, order=4)
            vals_r[:] = apply_filter(vals_r.values, b, a)
        elif name == "ACC":
            # construir magnitud si hay 3 ejes
            if vals_r.shape[1] == 3:
                mag = np.linalg.norm(vals_r.values, axis=1, keepdims=True)
                vals_r = pd.DataFrame(mag, index=vals_r.index, columns=["ACC_mag"])
            b, a = butter_lowpass(2.0, TARGET_FS, order=4)
            vals_r[:] = apply_filter(vals_r.values, b, a)
        ts[name] = vals_r
    # 2) alinear índices (inner join por tiempo)
    aligned = pd.concat(ts.values(), axis=1, join="inner")
    aligned.columns = list(ts.keys())
    return aligned
```

```python
def corr_matrix(df: pd.DataFrame) -> pd.DataFrame:
    return df.corr(method="pearson")
```

```python
def coherence_matrix(df: pd.DataFrame, fs=TARGET_FS, band=(0.1, 1.5)) -> pd.DataFrame:
    names = df.columns
    C = np.zeros((len(names), len(names)))
    for i, a in enumerate(names):
        for j, b in enumerate(names):
            if j < i: continue
            f, Pxy = sp.csd(df[a].values, df[b].values, fs=fs, nperseg=256)
            # integrar en banda
            mask = (f >= band[0]) & (f <= band[1])
            # coherencia normalizada: |Pxy|^2/(Pxx*Pyy)
            f_, Pxx = sp.welch(df[a].values, fs=fs, nperseg=256)
            f_, Pyy = sp.welch(df[b].values, fs=fs, nperseg=256)
            coh = (np.abs(Pxy[mask])**2) / (np.interp(f[mask], f_, Pxx) * np.interp(f[mask], f_, Pyy) + 1e-12)
            C[i, j] = C[j, i] = float(np.mean(coh))
    return pd.DataFrame(C, index=names, columns=names)
```

```python
def threshold_matrix(W: pd.DataFrame, sparsity=0.8) -> pd.DataFrame:
    # conserva top (1 - sparsity) de pesos absolutos
    A = W.copy()
    np.fill_diagonal(A.values, 0.0)
    vals = np.abs(A.values[np.triu_indices_from(A, k=1)])
    thr = np.quantile(vals, sparsity)
    A_thr = A.where(np.abs(A) >= thr, other=0.0)
    return A_thr
```

```python
def build_graph(W: pd.DataFrame, weighted=True) -> nx.Graph:
    G = nx.Graph()
    for i in W.index:
        G.add_node(i)
    for i in W.index:
        for j in W.columns:
            if i >= j: continue
            w = float(W.loc[i, j])
            if w != 0.0:
                G.add_edge(i, j, weight=w if weighted else 1.0)
    return G
```

```python
# Modelo nulo: aleatorización que preserva el grado
def randomize_graph_degree_preserving(G: nx.Graph, nswap=100, max_tries=1000):
    Gr = G.copy()
    nx.double_edge_swap(Gr, nswap=nswap, max_tries=max_tries, seed=42)
    return Gr
```

```python
def analyze_graph(G: nx.Graph):
    # Centralities
    deg = nx.degree_centrality(G)
    bet = nx.betweenness_centrality(G, normalized=True)
    clo = nx.closeness_centrality(G)
    eig = nx.eigenvector_centrality_numpy(G)
    # Clustering
    clust = nx.clustering(G, weight='weight')
    C_avg = float(np.mean(list(clust.values())))
    # Paths
    if nx.is_connected(G):
        L = nx.average_shortest_path_length(G, weight=None)
    else:
        L = np.nan
    # Rich-club
    rc = nx.rich_club_coefficient(G, normalized=False, Q=100)
    # Core-periphery: k-core number
    core_num = nx.core_number(G)
    # Motifs: triangles
    triangles = nx.triangles(G)
    # Communities: Louvain (python-louvain) no estándar en nx; usamos greedy modularity
    comms = list(nx.algorithms.community.greedy_modularity_communities(G, weight='weight'))
    part = {}
    for i, c in enumerate(comms):
        for n in c:
            part[n] = i
    return {
        "deg": deg, "bet": bet, "clo": clo, "eig": eig,
        "clust": clust, "C_avg": C_avg, "L": L,
        "rich_club": rc,
        "core_num": core_num,
        "triangles": triangles,
        "communities": part
    }
```

```python
def plot_heatmap(W: pd.DataFrame, title: str):
    plt.figure(figsize=(6,5))
    sns.heatmap(W, cmap="viridis", annot=False, square=True)
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
def plot_graph(G: nx.Graph, metrics: dict, title: str, dim3=False):
    # node size ~ degree, color ~ community
    deg = metrics["deg"]
    part = metrics["communities"]
    sizes = np.array([3000 * deg[n] for n in G.nodes()])
    colors = np.array([part[n] for n in G.nodes()])
    pos = nx.spring_layout(G, seed=42, dim=3 if dim3 else 2)
    plt.figure(figsize=(7,6))
    nx.draw_networkx(G, pos=pos, node_size=sizes, node_color=colors, cmap="tab10",
                     with_labels=True, width=[d['weight'] for _,_,d in G.edges(data=True)])
    plt.title(title)
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
def process_exam(subject_path: path, exam: str):
    exam_path = subject_path / exam
    files = {
        "EDA": exam_path / "EDA.csv",
        "BVP": exam_path / "BVP.csv",
        "HR":  exam_path / "HR.csv",
        "IBI": exam_path / "IBI.csv",
        "ACC": exam_path / "ACC.csv",
        "TEMP": exam_path / "TEMP.csv",
    }
    raw = {k: load_csv(v) for k, v in files.items() if v.exists()}
    # fs_guess puede leerse de info.txt (no imprescindible para este pipeline por uso de resample)
    fs_guess = {}
    df_aligned = preprocess_modalities(raw, fs_guess)

    # Conectividad: correlación y coherencia
    W_corr = corr_matrix(df_aligned)
    W_coh  = coherence_matrix(df_aligned, fs=TARGET_FS, band=(0.1, 1.5))

    # Umbralización (80% sparsidad => top 20% conexiones)
    Wc_thr = threshold_matrix(W_corr, sparsity=0.8)
    Wk_thr = threshold_matrix(W_coh,  sparsity=0.8)

    # Grafo y métricas
    Gc = build_graph(Wc_thr, weighted=True)
    Gk = build_graph(Wk_thr, weighted=True)
    metr_corr = analyze_graph(Gc)
    metr_coh  = analyze_graph(Gk)

    # Null: degree-preserving
    Gc_null = randomize_graph_degree_preserving(Gc, nswap=50, max_tries=500)
    Gk_null = randomize_graph_degree_preserving(Gk, nswap=50, max_tries=500)

    # Visualizaciones
    plot_heatmap(W_corr, f"Correlation connectivity – {subject_path.name} – {exam}")
    plot_heatmap(W_coh,  f"Coherence connectivity – {subject_path.name} – {exam}")
    plot_graph(Gc, metr_corr, f"Graph (corr) – {subject_path.name} – {exam}", dim3=False)
    plot_graph(Gk, metr_coh,  f"Graph (coh) – {subject_path.name} – {exam}", dim3=True)

    # Hubs por percentil 90 en grado/eigenvector
    deg_vals = pd.Series(metr_corr["deg"])
    eig_vals = pd.Series(metr_corr["eig"])
    hubs_deg = list(deg_vals[deg_vals >= deg_vals.quantile(0.9)].index)
    hubs_eig = list(eig_vals[eig_vals >= eig_vals.quantile(0.9)].index)

    return {
        "aligned": df_aligned,
        "W_corr": W_corr, "W_coh": W_coh,
        "Wc_thr": Wc_thr, "Wk_thr": Wk_thr,
        "Gc": Gc, "Gk": Gk,
        "metrics_corr": metr_corr, "metrics_coh": metr_coh,
        "hubs_deg": hubs_deg, "hubs_eig": hubs_eig,
        "Gc_null": Gc_null, "Gk_null": Gk_null
    }
```

```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
from mpl_toolkits.mplot3d import Axes3D

# Funciones auxiliares
def load_csv(path): return pd.read_csv(path, header=None)
def guess_unit(series): return "ms" if series.iloc[:100].median() > 1e11 else "s"
def to_timeseries(df, unit): 
    t = df.iloc[:,0]
    vals = df.iloc[:,1:].copy()
    vals.index = pd.to_datetime(t, unit=unit)
    return vals
def resample_to(df, fs=4.0):
    rule = pd.Timedelta(seconds=1.0/fs)
    return df.resample(rule).mean().interpolate(method="linear", limit=8)
def corr_matrix(df): return df.corr(method="pearson")
def threshold_matrix(W, sparsity=0.5):
    A = W.copy()
    np.fill_diagonal(A.values, 0.0)
    vals = np.abs(A.values[np.triu_indices_from(A, k=1)])
    thr = np.quantile(vals, sparsity)
    return A.where(np.abs(A) >= thr, other=0.0)
def build_graph(W):
    G = nx.Graph()
    for i in W.index:
        G.add_node(i)
    for i in W.index:
        for j in W.columns:
            if i != j and W.loc[i,j] != 0.0:
                G.add_edge(i, j, weight=W.loc[i,j])
    return G
def analyze_graph(G):
    deg = nx.degree_centrality(G)
    bet = nx.betweenness_centrality(G)
    clo = nx.closeness_centrality(G)
    eig = nx.eigenvector_centrality_numpy(G)
    clust = nx.clustering(G, weight='weight')
    C_avg = float(np.mean(list(clust.values())))
    L = nx.average_shortest_path_length(G) if nx.is_connected(G) else np.nan
    comms = list(nx.algorithms.community.greedy_modularity_communities(G, weight='weight'))
    part = {n: i for i, c in enumerate(comms) for n in c}
    return {"deg": deg, "bet": bet, "clo": clo, "eig": eig, "clust": clust, "C_avg": C_avg, "L": L, "communities": part}
```

```python
# Visualizaciones
def plot_heatmap(W, title):
    plt.figure(figsize=(6,5))
    sns.heatmap(W, cmap="viridis", annot=True, square=True)
    plt.title(title)
    plt.tight_layout()
    plt.show()

def plot_graph(G, metrics, title, dim3=False):
    deg = metrics["deg"]
    part = metrics["communities"]
    sizes = [3000 * deg[n] for n in G.nodes()]
    colors = [part[n] for n in G.nodes()]
    if not dim3:
        pos = nx.spring_layout(G, seed=42)
        plt.figure(figsize=(7,6))
        nx.draw_networkx(G, pos=pos, node_size=sizes, node_color=colors, cmap="tab10",
                         with_labels=True, width=[d['weight'] for _,_,d in G.edges(data=True)])
        plt.title(title)
        plt.axis('off')
        plt.tight_layout()
        plt.show()
    else:
        fig = plt.figure(figsize=(8,6))
        ax = fig.add_subplot(111, projection='3d')
        pos_3d = {n: np.random.rand(3) for n in G.nodes()}
        for n in G.nodes():
            ax.scatter(*pos_3d[n], s=3000*deg[n], c=colors[G.nodes().index(n)], cmap='tab10', edgecolors='k')
        for i,j in G.edges():
            x = [pos_3d[i][0], pos_3d[j][0]]
            y = [pos_3d[i][1], pos_3d[j][1]]
            z = [pos_3d[i][2], pos_3d[j][2]]
            ax.plot(x, y, z, c='gray', alpha=0.6)
        ax.set_title(title)
        plt.tight_layout()
        plt.show()
```
